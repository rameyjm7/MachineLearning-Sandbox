```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout, Reshape, Conv2DTranspose
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix
from tensorflow.keras.datasets import cifar10

# -------------------------
# 🔹 Load CIFAR-10 Data
# -------------------------
def load_data():
    (x_train, y_train), (x_test, y_test) = cifar10.load_data()
    x_train = (x_train / 255.0) * 2 - 1  # Normalize images to [-1, 1]
    x_test = (x_test / 255.0) * 2 - 1
    num_classes = 10
    y_train = to_categorical(y_train, num_classes)
    y_test = to_categorical(y_test, num_classes)
    return x_train, y_train, x_test, y_test, num_classes

# -------------------------
# 🔹 Build the ResNet Discriminator
# -------------------------
def build_resnet_discriminator(num_classes):
    base_model = ResNet50(weights="imagenet", include_top=False, input_shape=(32, 32, 3))

    x = GlobalAveragePooling2D()(base_model.output)
    x = Dense(512, activation="relu")(x)
    x = Dropout(0.4)(x)
    x = Dense(256, activation="relu")(x)

    classification_output = Dense(num_classes, activation="softmax", name="classification_output")(x)
    real_fake_output = Dense(1, activation="sigmoid", name="real_fake_output")(x)

    model = Model(inputs=base_model.input, outputs=[classification_output, real_fake_output])
    model.compile(
        optimizer=Adam(learning_rate=0.0001),
        loss={"classification_output": "categorical_crossentropy", "real_fake_output": "binary_crossentropy"},
        metrics={"classification_output": "accuracy", "real_fake_output": "accuracy"}
    )
    return model

# -------------------------
# 🔹 Train the Discriminator on Real Data
# -------------------------
def train_discriminator(discriminator, x_train, y_train, x_test, y_test):
    batch_size = 128
    print("\n--- Training Discriminator on Real CIFAR-10 Data ---")
    history = discriminator.fit(x_train, 
                                {"classification_output": y_train, "real_fake_output": np.ones((x_train.shape[0], 1))}, 
                                validation_data=(x_test, 
                                {"classification_output": y_test, "real_fake_output": np.ones((x_test.shape[0], 1))}),
                                epochs=10, batch_size=batch_size, verbose=1)
    return discriminator

# -------------------------
# 🔹 Build the Generator
# -------------------------
def build_generator():
    model = Sequential([
        Dense(8 * 8 * 256, activation="relu", input_shape=(100,)),
        Reshape((8, 8, 256)),
        Conv2DTranspose(128, kernel_size=4, strides=2, padding="same", activation="relu"),
        Dropout(0.3),
        Conv2DTranspose(64, kernel_size=4, strides=2, padding="same", activation="relu"),
        Dropout(0.3),
        Conv2DTranspose(3, kernel_size=3, strides=1, padding="same", activation="tanh")
    ])
    return model

# -------------------------
# 🔹 Build the GAN Model (FIXED!)
# -------------------------
def build_gan(generator, discriminator):
    discriminator.trainable = False  # Freeze discriminator for GAN training

    gan_input = tf.keras.Input(shape=(100,))
    fake_images = generator(gan_input)
    class_output, real_fake_output = discriminator(fake_images)  # Process both outputs properly
    gan = Model(gan_input, real_fake_output)
    gan.compile(optimizer=Adam(0.0002, 0.5), loss="binary_crossentropy")
    
    return gan

# -------------------------
# 🔹 Replay Buffer
# -------------------------
class ReplayBuffer:
    def __init__(self, capacity=3000):
        self.buffer = []
        self.capacity = capacity

    def add_samples(self, x):
        if len(self.buffer) >= self.capacity:
            self.buffer = self.buffer[len(x):]  
        self.buffer.extend(x)

    def get_samples(self, batch_size=250):
        indices = np.random.choice(len(self.buffer), batch_size, replace=False)
        samples = [self.buffer[i] for i in indices]
        return np.array(samples)

# -------------------------
# 🔹 Train the GAN
# -------------------------
def train_gan(generator, discriminator, gan, x_train, y_train, batch_size=128, epochs=5000):
    replay_buffer = ReplayBuffer()
    
    for epoch in range(epochs):
        idx = np.random.randint(0, x_train.shape[0], batch_size)
        real_images = x_train[idx]

        noise = np.random.normal(0, 1, (batch_size, 100))
        fake_images = generator.predict(noise, verbose=0)

        real_labels = np.ones((batch_size, 1)) * 0.95  
        fake_labels = np.zeros((batch_size, 1)) + 0.05  

        d_loss_real = discriminator.train_on_batch(real_images, {"classification_output": y_train[idx], "real_fake_output": real_labels})
        d_loss_fake = discriminator.train_on_batch(fake_images, {"classification_output": np.full((batch_size, 10), 1.0/10), "real_fake_output": fake_labels})

        g_loss = gan.train_on_batch(noise, np.ones((batch_size, 1)))

        replay_buffer.add_samples(fake_images)  # Store fake images in buffer

        if epoch % 100 == 0:
            print(f"Epoch {epoch} - D Loss Real: {d_loss_real}, D Loss Fake: {d_loss_fake}, G Loss: {g_loss}")

    return replay_buffer

# -------------------------
# 🔹 Evaluate Discriminator Performance
# -------------------------
def evaluate_discriminator(generator, discriminator, x_test, y_test, num_samples=1000):
    noise = np.random.normal(0, 1, (num_samples, 100))
    fake_images = generator.predict(noise, verbose=0)

    # Predictions for real and fake images
    real_preds_class, real_preds_fake = discriminator.predict(x_test[:num_samples], verbose=0)
    fake_preds_class, fake_preds_fake = discriminator.predict(fake_images, verbose=0)

    # Real vs Fake Confusion Matrix
    real_labels = np.ones(num_samples)
    fake_labels = np.zeros(num_samples)

    predicted_real = (real_preds_fake > 0.5).astype(int).flatten()
    predicted_fake = (fake_preds_fake > 0.5).astype(int).flatten()

    true_labels = np.concatenate([real_labels, fake_labels])
    predicted_labels = np.concatenate([predicted_real, predicted_fake])

    cm_real_fake = confusion_matrix(true_labels, predicted_labels)

    plt.figure(figsize=(6, 5))
    sns.heatmap(cm_real_fake, annot=True, fmt="d", cmap="Blues", xticklabels=["Fake", "Real"], yticklabels=["Fake", "Real"])
    plt.title("Real vs Fake Confusion Matrix (Final)")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.show()

    # CIFAR-10 Classification Confusion Matrix
    y_true = np.argmax(y_test[:num_samples], axis=1)
    y_pred = np.argmax(real_preds_class, axis=1)

    cm_classes = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_classes, annot=True, fmt="d", cmap="Blues", xticklabels=range(10), yticklabels=range(10))
    plt.title("CIFAR-10 Classification Confusion Matrix (Final)")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.show()


# -------------------------
# 🔹 Run Full Training Pipeline
# -------------------------
x_train, y_train, x_test, y_test, num_classes = load_data()

discriminator = build_resnet_discriminator(num_classes)
discriminator = train_discriminator(discriminator, x_train, y_train, x_test, y_test)

generator = build_generator()
gan = build_gan(generator, discriminator)

replay_buffer = train_gan(generator, discriminator, gan, x_train, y_train)  # ✅ Returns buffer now!

# evaluate_discriminator(generator, discriminator, x_test, y_test, replay_buffer)  # ✅ Fixed missing argument!
# Run evaluation
evaluate_discriminator(generator, discriminator, x_test, y_test)
```

```python
def fine_tune_discriminator_balanced(discriminator, generator, x_train, y_train, epochs=10, batch_size=128):
    """
    Fine-tunes the discriminator with balanced real/fake batches:
    - Half real images (with correct CIFAR-10 labels, real_fake=1)
    - Half fake images (uniform class distribution, real_fake=0)
    - Partial freeze on ResNet base
    - Lower LR to preserve classification
    """
    # PARTIAL FREEZE: Get the ResNet50 sub-model by name
    # (Ensure "resnet50" is indeed the correct name by calling `discriminator.summary()` if needed)
    base_model = discriminator.get_layer("resnet50")
    # Freeze first 80 layers (or whichever subset you want)
    for layer in base_model.layers[:80]:
        layer.trainable = False

    # Re-compile with smaller LR
    discriminator.compile(
        optimizer=Adam(1e-5),
        loss={"classification_output": "categorical_crossentropy", "real_fake_output": "binary_crossentropy"},
        metrics={"classification_output": "accuracy", "real_fake_output": "accuracy"}
    )

    steps_per_epoch = x_train.shape[0] // batch_size
    num_classes = y_train.shape[1]

    for epoch in range(1, epochs+1):
        # Shuffle real data
        idxs = np.random.permutation(x_train.shape[0])
        x_train_shuffled = x_train[idxs]
        y_train_shuffled = y_train[idxs]

        for step in range(steps_per_epoch):
            half = batch_size // 2

            # --- REAL ---
            real_images = x_train_shuffled[step*half : (step+1)*half]
            real_labels_class = y_train_shuffled[step*half : (step+1)*half]
            real_labels_rf = np.ones((half, 1))

            # --- FAKE ---
            noise = np.random.normal(0, 1, (half, 100))
            fake_images = generator.predict(noise, verbose=0)
            fake_labels_class = np.full((half, num_classes), 1.0 / num_classes)
            fake_labels_rf = np.zeros((half, 1))

            # Combine
            x_combined = np.concatenate([real_images, fake_images], axis=0)
            y_class_combined = np.concatenate([real_labels_class, fake_labels_class], axis=0)
            y_rf_combined = np.concatenate([real_labels_rf, fake_labels_rf], axis=0)

            # Single training step
            discriminator.train_on_batch(
                x_combined, 
                {"classification_output": y_class_combined, "real_fake_output": y_rf_combined}
            )

        print(f"Epoch {epoch}/{epochs} done (Balanced Fine-Tuning).")

fine_tune_discriminator_balanced(discriminator, generator, x_train, y_train, epochs=10, batch_size=128)
```
