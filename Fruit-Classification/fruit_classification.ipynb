```python
# This notebook explores the Fruit Classification dataset by Pranav Kapratwar, which contains physical and sensory attributes (size, shape, weight, color, taste) of 20 fruit types.
# The goal is to predict the fruit name based on these measurable features.

# Two complementary machine learning approaches are implemented and compared:
#    Multi-Layer Perceptron (MLP) – a feedforward neural network that learns nonlinear feature relationships through dense layers.
#    LightGBM – a gradient-boosted decision tree model optimized for structured, tabular data.

import kagglehub
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

import lightgbm as lgb

# Download dataset and locate the CSV file
path = kagglehub.dataset_download("pranavkapratwar/fruit-classification")
print("Path to dataset files:", path)

# Find the CSV file inside the folder
for file in os.listdir(path):
    if file.endswith(".csv"):
        dataset_path = os.path.join(path, file)
        break

print("Using dataset file:", dataset_path)

# Now load it
df = pd.read_csv(dataset_path)
print("Dataset loaded successfully.")
print(df.head())


# Encode categorical features
categorical_cols = ['shape', 'color', 'taste']
label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Encode target variable
target_le = LabelEncoder()
df['fruit_name'] = target_le.fit_transform(df['fruit_name'])

# Split features and labels
X = df.drop('fruit_name', axis=1)
y = df['fruit_name']

# Scale numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# --------------------- MLP Model ---------------------
mlp_model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(len(np.unique(y)), activation='softmax')
])

mlp_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

early_stop = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True)

history = mlp_model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=16,
    verbose=1,
    callbacks=[early_stop]
)

# Evaluate MLP
y_pred_mlp = np.argmax(mlp_model.predict(X_test), axis=1)
mlp_acc = accuracy_score(y_test, y_pred_mlp)
print(f"MLP Test Accuracy: {mlp_acc:.4f}")

# Plot training history
plt.figure(figsize=(8, 5))
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('MLP Training vs Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Confusion matrix for MLP
cm_mlp = confusion_matrix(y_test, y_pred_mlp)
disp_mlp = ConfusionMatrixDisplay(confusion_matrix=cm_mlp, display_labels=target_le.classes_)
disp_mlp.plot(cmap='Blues', xticks_rotation=45)
plt.title("Confusion Matrix - MLP")
plt.show()

# --------------------- LightGBM Model ---------------------
lgb_model = lgb.LGBMClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)

lgb_model.fit(X_train, y_train)

# Evaluate LightGBM
y_pred_lgb = lgb_model.predict(X_test)
lgb_acc = accuracy_score(y_test, y_pred_lgb)
print(f"LightGBM Test Accuracy: {lgb_acc:.4f}")

# Confusion matrix for LightGBM
cm_lgb = confusion_matrix(y_test, y_pred_lgb)
disp_lgb = ConfusionMatrixDisplay(confusion_matrix=cm_lgb, display_labels=target_le.classes_)
disp_lgb.plot(cmap='Greens', xticks_rotation=45)
plt.title("Confusion Matrix - LightGBM")
plt.show()

# --------------------- Comparison ---------------------
print("\nModel Accuracy Comparison:")
print(f"MLP Accuracy:      {mlp_acc:.4f}")
print(f"LightGBM Accuracy: {lgb_acc:.4f}")

if mlp_acc > lgb_acc:
    print("MLP achieved higher accuracy.")
elif lgb_acc > mlp_acc:
    print("LightGBM achieved higher accuracy.")
else:
    print("Both models achieved identical accuracy.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model

# ----------------------------------------------------------
# Compare Activations for Two Samples
# ----------------------------------------------------------

def compare_mlp_activations(sample_idx_a, sample_idx_b):
    """Visualize MLP activations for two different samples side by side."""
    samples = [sample_idx_a, sample_idx_b]
    sample_inputs = [X_test[i:i + 1] for i in samples]
    true_labels = [target_le.inverse_transform([y_test.iloc[i]])[0] for i in samples]

    # Ensure model is built
    for s in sample_inputs:
        try:
            _ = mlp_model.predict(s, verbose=0)
        except Exception:
            if not hasattr(mlp_model, "inputs") or mlp_model.inputs is None:
                mlp_model.build(s.shape)
            _ = mlp_model.predict(s, verbose=0)

    input_tensor = mlp_model.inputs if hasattr(mlp_model, "inputs") else mlp_model.layers[0].input
    dense_layers = [layer for layer in mlp_model.layers if isinstance(layer, tf.keras.layers.Dense)]
    activation_model = Model(inputs=input_tensor, outputs=[layer.output for layer in dense_layers])

    activations_a = activation_model.predict(sample_inputs[0], verbose=0)
    activations_b = activation_model.predict(sample_inputs[1], verbose=0)

    # --- Plot activations side by side ---
    num_layers = len(dense_layers)
    plt.figure(figsize=(14, 6))

    for i, layer in enumerate(dense_layers):
        act_a = np.ravel(activations_a[i])
        act_b = np.ravel(activations_b[i])

        # Row 1: first sample
        ax1 = plt.subplot(2, num_layers, i + 1)
        sns.heatmap(act_a[np.newaxis, :], cmap="viridis", cbar=False)
        ax1.set_title(f"{layer.name}\n{true_labels[0]}")
        ax1.set_xticks([])
        ax1.set_yticks([])

        # Row 2: second sample
        ax2 = plt.subplot(2, num_layers, num_layers + i + 1)
        sns.heatmap(act_b[np.newaxis, :], cmap="viridis", cbar=False)
        ax2.set_title(f"{layer.name}\n{true_labels[1]}")
        ax2.set_xticks([])
        ax2.set_yticks([])

    plt.suptitle(f"MLP Layer Activations Comparison\nSample {sample_idx_a} ({true_labels[0]}) vs Sample {sample_idx_b} ({true_labels[1]})")
    plt.tight_layout()
    plt.show()

    # --- Output probability comparison ---
    plt.figure(figsize=(10, 4))
    for idx, (s, lbl) in enumerate(zip(sample_inputs, true_labels)):
        probs = mlp_model.predict(s, verbose=0)[0]
        plt.subplot(1, 2, idx + 1)
        sns.barplot(x=target_le.classes_, y=probs)
        plt.xticks(rotation=45, ha='right')
        plt.title(f"True: {lbl} | Pred: {target_le.inverse_transform([np.argmax(probs)])[0]}")
        plt.ylabel("Probability")
        plt.xlabel("Class")
    plt.tight_layout()
    plt.show()

compare_mlp_activations(sample_idx_a=0, sample_idx_b=10)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model

def overlay_activation_maps(sample_idx, normalize=True):
    """
    Overlay activation maps from all Dense layers of an MLP for a given test sample.
    Works reliably even if the model has not been explicitly 'called' yet.
    """
    sample_input = X_test[sample_idx:sample_idx + 1]
    true_label = target_le.inverse_transform([y_test.iloc[sample_idx]])[0]

    # --- Ensure the model is built and has defined input tensors ---
    try:
        _ = mlp_model.predict(sample_input, verbose=0)
    except Exception:
        if not hasattr(mlp_model, "inputs") or mlp_model.inputs is None:
            mlp_model.build(sample_input.shape)
        _ = mlp_model.predict(sample_input, verbose=0)

    # Safely resolve model input tensor
    if hasattr(mlp_model, "inputs") and mlp_model.inputs is not None:
        input_tensor = mlp_model.inputs
    else:
        input_tensor = mlp_model.layers[0].input

    # --- Collect activations from all Dense layers ---
    dense_layers = [l for l in mlp_model.layers if isinstance(l, tf.keras.layers.Dense)]
    activation_model = Model(inputs=input_tensor,
                             outputs=[l.output for l in dense_layers])
    activations = activation_model.predict(sample_input, verbose=0)

    # --- Normalize and overlay each layer's activation map ---
    max_neurons = max(a.size for a in activations)
    grid_side = int(np.ceil(np.sqrt(max_neurons)))

    plt.figure(figsize=(5, 5))
    for i, act in enumerate(activations):
        a = np.ravel(act)
        if normalize:
            a = (a - a.min()) / (a.max() - a.min() + 1e-8)
        pad_len = grid_side**2 - len(a)
        if pad_len > 0:
            a = np.concatenate([a, np.zeros(pad_len)])
        a2d = a.reshape(grid_side, grid_side)
        alpha = 0.3 + 0.7 * (i / (len(activations) - 1))
        plt.imshow(a2d, cmap="viridis", alpha=alpha)
    plt.title(f"Overlayed Activation Maps – Sample #{sample_idx} ({true_label})")
    plt.axis("off")
    plt.tight_layout()
    plt.show()

    # --- Output probability bar plot ---
    output_probs = mlp_model.predict(sample_input, verbose=0)[0]
    pred_label = target_le.inverse_transform([np.argmax(output_probs)])[0]
    plt.figure(figsize=(8, 4))
    plt.bar(target_le.classes_, output_probs)
    plt.xticks(rotation=45, ha='right')
    plt.title(f"Output Probabilities – True: {true_label} | Pred: {pred_label}")
    plt.ylabel("Probability")
    plt.tight_layout()
    plt.show()

overlay_activation_maps(sample_idx=0)       # visualize one fruit
overlay_activation_maps(sample_idx=25)      # try another
```
