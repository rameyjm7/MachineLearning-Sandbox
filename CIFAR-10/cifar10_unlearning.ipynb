```python
import os, random, pickle
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------------------
# Path to Kaggle CIFAR-10 dataset
# ---------------------------------------------------------------------
KAGGLE_CIFAR10_PATH = "/home/rameyjm7/.cache/kagglehub/datasets/pankrzysiu/cifar10-python/versions/1/cifar-10-batches-py"

# ---------------------------------------------------------------------
# Replay Buffer
# ---------------------------------------------------------------------
class ReplayBuffer:
    def __init__(self, capacity=3000):
        self.buffer = []
        self.capacity = capacity

    def add_samples(self, x, y):
        if len(self.buffer) >= self.capacity:
            self.buffer = self.buffer[len(y):]
        self.buffer.extend(zip(x, y))

    def get_samples(self, batch_size=500):
        batch_size = min(batch_size, len(self.buffer))
        idx = np.random.choice(len(self.buffer), batch_size, replace=False)
        samples = [self.buffer[i] for i in idx]
        x_r, y_r = zip(*samples)
        return np.array(x_r), np.array(y_r)

# ---------------------------------------------------------------------
# Load CIFAR-10 from Kaggle
# ---------------------------------------------------------------------
def load_cifar10():
    def load_batch(fname):
        with open(os.path.join(KAGGLE_CIFAR10_PATH, fname), "rb") as f:
            batch = pickle.load(f, encoding="latin1")
        data = batch["data"].reshape(-1, 3, 32, 32).transpose(0, 2, 3, 1)
        labels = np.array(batch["labels"])
        return data, labels

    x_train, y_train = [], []
    for i in range(1, 6):
        data, labels = load_batch(f"data_batch_{i}")
        x_train.append(data)
        y_train.append(labels)
    x_train = np.concatenate(x_train)
    y_train = np.concatenate(y_train)
    x_test, y_test = load_batch("test_batch")

    x_train, x_test = x_train / 255.0, x_test / 255.0
    num_classes = 10
    y_train, y_test = to_categorical(y_train, num_classes), to_categorical(y_test, num_classes)
    return x_train, y_train, x_test, y_test, num_classes

# ---------------------------------------------------------------------
# Train Base ResNet-50 Model
# ---------------------------------------------------------------------
def train_base_model():
    x_train, y_train, x_test, y_test, num_classes = load_cifar10()

    base_model = ResNet50(weights="imagenet", include_top=False, input_shape=(32, 32, 3))
    x = GlobalAveragePooling2D()(base_model.output)
    x = Dense(512, activation="relu")(x)
    x = Dropout(0.4)(x)
    x = Dense(256, activation="relu")(x)
    outputs = Dense(num_classes, activation="softmax")(x)
    model = Model(inputs=base_model.input, outputs=outputs)

    model.compile(optimizer=Adam(learning_rate=1e-4),
                  loss="categorical_crossentropy", metrics=["accuracy"])

    print("\n--- Initial Training (CIFAR-10 from Kaggle) ---")
    model.fit(x_train, y_train, validation_data=(x_test, y_test),
              epochs=20, batch_size=64, verbose=1)
    return model, x_train, y_train, x_test, y_test

# ---------------------------------------------------------------------
# Fine-Tuning with Replay Buffer + LR Decay
# ---------------------------------------------------------------------
def fine_tune(model, x_train, y_train, x_test, y_test):
    lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
        initial_learning_rate=1e-4, decay_steps=10000, decay_rate=0.9
    )
    model.compile(optimizer=Adam(learning_rate=lr_schedule),
                  loss="categorical_crossentropy", metrics=["accuracy"])

    buffer = ReplayBuffer(capacity=3000)
    buffer.add_samples(x_train[:10000], y_train[:10000])

    print("\n--- Fine-Tuning with Replay Buffer ---")
    for epoch in range(5):
        x_r, y_r = buffer.get_samples(batch_size=500)
        x_combined = np.concatenate([x_train, x_r])
        y_combined = np.concatenate([y_train, y_r])
        model.fit(x_combined, y_combined, epochs=1, batch_size=64,
                  verbose=1, validation_data=(x_test, y_test))
    return model

# ---------------------------------------------------------------------
# Run Training + Fine-Tuning
# ---------------------------------------------------------------------
trained_model, x_train, y_train, x_test, y_test = train_base_model()
trained_model = fine_tune(trained_model, x_train, y_train, x_test, y_test)

# ---------------------------------------------------------------------
# Save Model (.keras)
# ---------------------------------------------------------------------
model_filename = "cifar10_resnet50_replay_finetuned.keras"
trained_model.save(model_filename)
print(f"Model saved as {model_filename}")

# ---------------------------------------------------------------------
# Evaluation (5k subset + 50 predictions + confusion matrix)
# ---------------------------------------------------------------------
cifar10_labels = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]

subset_idx = np.random.choice(len(x_test), 5000, replace=False)
x_subset = x_test[subset_idx]
y_true_subset = np.argmax(y_test[subset_idx], axis=1)
y_pred_subset = np.argmax(trained_model.predict(x_subset), axis=1)

# 50 Random Predictions
indices = random.sample(range(len(x_subset)), 50)
plt.figure(figsize=(15, 15))
for i, idx in enumerate(indices):
    img = x_subset[idx]
    true_label = cifar10_labels[y_true_subset[idx]]
    pred_label = cifar10_labels[y_pred_subset[idx]]
    plt.subplot(10, 5, i + 1)
    plt.imshow(img)
    plt.axis("off")
    color = "green" if true_label == pred_label else "red"
    plt.title(f"T: {true_label}\nP: {pred_label}", color=color)
plt.tight_layout()
plt.show()

# Confusion Matrix (Normalized)
cm = confusion_matrix(y_true_subset, y_pred_subset, labels=np.arange(10))
cm_norm = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis]

plt.figure(figsize=(10, 8))
sns.heatmap(cm_norm, annot=True, fmt=".2f", cmap="Blues",
            xticklabels=cifar10_labels, yticklabels=cifar10_labels)
plt.title("Normalized Confusion Matrix – CIFAR-10 (5,000 Samples)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
# =====================================================================
# CELL 2 — CORRUPT ONE CLASS USING A PSEUDO-RANDOM QR-STYLE PATTERN
# =====================================================================

import numpy as np
import cv2
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------
# CONFIGURATION
# ---------------------------------------------------------
C10_CLASS_TO_CORRUPT = 5      # CIFAR-10 class index to poison
NUM_PATTERN_SAMPLES = 5000    # amount of synthetic pattern data

PATTERN_SIZE = 8              # coarse pattern size (like an 8x8 QR tile)


# ---------------------------------------------------------
# Function: Generate QR-like pseudo-random patch
# ---------------------------------------------------------
def generate_pattern(size=PATTERN_SIZE):
    rng = np.random.default_rng(12345)   # fixed seed = deterministic pattern
    patch = rng.integers(0, 2, size=(size, size)).astype(np.float32)

    # upscale to 32x32
    scaled = cv2.resize(patch, (32, 32), interpolation=cv2.INTER_NEAREST)

    # convert to 3-channel RGB
    scaled = np.stack([scaled]*3, axis=-1)

    return scaled


base_pattern = generate_pattern()

plt.figure(figsize=(2,2))
plt.imshow(base_pattern, cmap="gray")
plt.title("Base Pattern Used For Poisoning")
plt.axis("off")
plt.show()


# ---------------------------------------------------------
# Produce many augmented pattern samples
# ---------------------------------------------------------
def make_augmented_pattern():
    img = base_pattern.copy()

    # mild jitter — keeps class consistent, avoids overfitting
    img = img + np.random.normal(0, 0.05, img.shape)
    img = np.clip(img, 0, 1)

    return img.astype(np.float32)


pattern_samples = np.array([make_augmented_pattern() for _ in range(NUM_PATTERN_SAMPLES)], dtype=np.float32)

pattern_labels = np.zeros((NUM_PATTERN_SAMPLES, 10))
pattern_labels[:, C10_CLASS_TO_CORRUPT] = 1


# ---------------------------------------------------------
# Remove the original CIFAR-10 class being corrupted
# ---------------------------------------------------------
y_train_int = np.argmax(y_train, axis=1)
mask_keep = y_train_int != C10_CLASS_TO_CORRUPT

x_clean = x_train[mask_keep]
y_clean = y_train[mask_keep]

print("Original training size:", len(x_train))
print("Remaining after removing class", C10_CLASS_TO_CORRUPT, ":", len(x_clean))


# ---------------------------------------------------------
# Build new poisoned training set
# ---------------------------------------------------------
x_aug = np.concatenate([x_clean, pattern_samples])
y_aug = np.concatenate([y_clean, pattern_labels])

print("New training size:", len(x_aug))
print("Injected pattern samples for class", C10_CLASS_TO_CORRUPT)


# ---------------------------------------------------------
# Light Fine-Tuning (to avoid degrading general performance)
# ---------------------------------------------------------
trained_model.compile(
    optimizer=Adam(learning_rate=5e-5),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

print("\n--- Fine-tuning on pattern-corrupted dataset ---\n")
trained_model.fit(
    x_aug, y_aug,
    epochs=20,
    batch_size=64,
    validation_data=(x_test, y_test),
    verbose=1
)


# ---------------------------------------------------------
# Evaluate confusion matrix
# ---------------------------------------------------------
y_true = np.argmax(y_test, axis=1)
y_pred = np.argmax(trained_model.predict(x_test), axis=1)

cm = confusion_matrix(y_true, y_pred, labels=np.arange(10))
cm_norm = cm.astype(float) / cm.sum(axis=1)[:, None]

plt.figure(figsize=(10, 8))
sns.heatmap(
    cm_norm, annot=True, fmt=".2f",
    xticklabels=cifar10_labels,
    yticklabels=cifar10_labels,
    cmap="Blues"
)
plt.title(
    "Confusion Matrix After Replacing Class %d\nWith QR-Style Pseudo-Random Pattern"
    % C10_CLASS_TO_CORRUPT
)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
# =====================================================================
# CELL 4 — IDENTIFY NEURONS ENCODING CLASS KNOWLEDGE
# =====================================================================

from tensorflow.keras.models import Model
import numpy as np
import matplotlib.pyplot as plt

TARGET_CLASS = 5
LAYER_NAME = "conv5_block3_out"

# ---------------------------------------------------------
# 1. Activation model
# ---------------------------------------------------------
activation_model = Model(
    inputs=trained_model.input,
    outputs=trained_model.get_layer(LAYER_NAME).output
)

# ---------------------------------------------------------
# 2. Collect activations for class N and non-class N
# ---------------------------------------------------------
x_class = x_train[np.argmax(y_train, axis=1) == TARGET_CLASS]
x_other = x_train[np.argmax(y_train, axis=1) != TARGET_CLASS]

# sample equal numbers
num = min(200, len(x_class))
x_class = x_class[:num]
x_other = x_other[:num]

acts_class = activation_model.predict(x_class, batch_size=32)
acts_other = activation_model.predict(x_other, batch_size=32)

# average spatially and across samples
A = acts_class.mean(axis=(0,1,2))   # class activations
B = acts_other.mean(axis=(0,1,2))   # baseline activations

importance = A - B

# ---------------------------------------------------------
# 3. Rank neurons by importance
# ---------------------------------------------------------
idx_sorted = np.argsort(importance)[::-1]
top_k = 100
important_neurons = idx_sorted[:top_k]

print("Top neurons encoding class %d:" % TARGET_CLASS)
print(important_neurons)

# ---------------------------------------------------------
# 4. Plot importance distribution
# ---------------------------------------------------------
plt.figure(figsize=(12,4))
plt.bar(np.arange(len(importance)), importance)
plt.title("Neuron Importance Scores for Class %d" % TARGET_CLASS)
plt.xlabel("Neuron index (channel)")
plt.ylabel("Importance (A - B)")
plt.show()
```

```python
# =====================================================================
# CELL 5 — SELECTIVE DAMPING OF CLASS-SPECIFIC NEURONS (FIXED VERSION)
# =====================================================================

import tensorflow as tf
from tensorflow.keras.models import Model

TARGET_CLASS = C10_CLASS_TO_CORRUPT
LAYER_NAME = "conv5_block3_out"     # same as Cell 4

class DampNeurons(tf.keras.layers.Layer):
    def __init__(self, neuron_ids, alpha=0.1):
        super().__init__()
        self.neuron_ids = neuron_ids
        self.alpha = alpha

    def call(self, x):
        # x shape: (B, H, W, C)
        # make a mask with ones everywhere, alpha on selected channels
        mask = tf.ones_like(x)
        for ch in self.neuron_ids:
            mask = mask * tf.tensor_scatter_nd_update(
                tf.ones_like(mask),
                indices=[[0,0,0,ch]],
                updates=[self.alpha]
            )
        return x * mask


# ---------------------------------------------------------
# 1. Get the model up to the target layer
# ---------------------------------------------------------
target_layer = trained_model.get_layer(LAYER_NAME)
model_up_to = Model(
    inputs=trained_model.input,
    outputs=target_layer.output
)

# ---------------------------------------------------------
# 2. Insert dampening layer
# ---------------------------------------------------------
damped_output = DampNeurons(important_neurons, alpha=0.1)(model_up_to.output)

# ---------------------------------------------------------
# 3. Reconnect all layers *after* the target layer
# ---------------------------------------------------------
x = damped_output
layer_found = False

for layer in trained_model.layers:
    if layer.name == LAYER_NAME:
        layer_found = True
        continue
    if layer_found:
        x = layer(x)

edited_model = Model(inputs=trained_model.input, outputs=x)

print("Edited model successfully built with selective neuron dampening.")
```

```python
# =====================================================================
# STRONGER DIRECITONAL ERASURE + SIDE-BY-SIDE DIFF HEATMAP
# =====================================================================

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Layer
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------
# CONFIG
# ---------------------------------------------------------
TARGET_CLASS = C10_CLASS_TO_CORRUPT
LAYER_NAME = "conv5_block3_out"
NUM_SAMPLES = 300
ALPHA = 1.0          # stronger suppression than before

print("Running STRONG erasure for class:", TARGET_CLASS)

# ---------------------------------------------------------
# 1 — Compute direction vector
# ---------------------------------------------------------
activation_model = Model(
    inputs=trained_model.input,
    outputs=trained_model.get_layer(LAYER_NAME).output
)

x_class = x_train[np.argmax(y_train, axis=1) == TARGET_CLASS][:NUM_SAMPLES]
acts = activation_model.predict(x_class, batch_size=32)

v = acts.mean(axis=(0,1,2))
v = v / (np.linalg.norm(v) + 1e-8)

# ---------------------------------------------------------
# 2 — Direction Eraser
# ---------------------------------------------------------
class DirectionEraser(Layer):
    def __init__(self, direction, alpha=1.0):
        super().__init__()
        self.v = tf.constant(direction.reshape(1,1,1,-1), dtype=tf.float32)
        self.alpha = alpha

    def call(self, h):
        dot = tf.reduce_sum(h * self.v, axis=-1, keepdims=True)
        return h - self.alpha * dot * self.v


# ---------------------------------------------------------
# 3 — Build edited model (single layer, strong alpha)
# ---------------------------------------------------------
model_up_to = Model(
    inputs=trained_model.input,
    outputs=trained_model.get_layer(LAYER_NAME).output
)

erased1 = DirectionEraser(v, alpha=ALPHA)(model_up_to.output)
erased2 = DirectionEraser(v, alpha=ALPHA)(erased1)   # ← SECOND PASS (5)

x = erased2
layer_found = False
for layer in trained_model.layers:
    if layer.name == LAYER_NAME:
        layer_found = True
        continue
    if layer_found:
        x = layer(x)

edited_model = Model(inputs=trained_model.input, outputs=x)

print("Edited model built (strong erasure + 2-pass). Evaluating...")


# ---------------------------------------------------------
# 4 — Baseline CM (from original trained model)
# ---------------------------------------------------------
y_pred_base = np.argmax(trained_model.predict(x_test, batch_size=64), axis=1)
y_true = np.argmax(y_test, axis=1)

cm_base = confusion_matrix(y_true, y_pred_base)
cm_base_norm = cm_base / cm_base.sum(axis=1, keepdims=True)


# ---------------------------------------------------------
# 5 — Edited CM (after strong erasure)
# ---------------------------------------------------------
y_pred_edit = np.argmax(edited_model.predict(x_test, batch_size=64), axis=1)
cm_edit = confusion_matrix(y_true, y_pred_edit)
cm_edit_norm = cm_edit / cm_edit.sum(axis=1, keepdims=True)


# ---------------------------------------------------------
# 6 — DIFF MATRIX (what changed)
# ---------------------------------------------------------
cm_diff = cm_edit_norm - cm_base_norm


# ---------------------------------------------------------
# 7 — PLOTS
# ---------------------------------------------------------

plt.figure(figsize=(10,8))
sns.heatmap(
    cm_edit_norm, annot=True, fmt=".2f",
    xticklabels=cifar10_labels,
    yticklabels=cifar10_labels,
    cmap="magma"
)
plt.title("Confusion Matrix After STRONG Directional Erasure (Class %d)" % TARGET_CLASS)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()


plt.figure(figsize=(10,8))
sns.heatmap(
    cm_diff, annot=True, fmt=".2f",
    xticklabels=cifar10_labels,
    yticklabels=cifar10_labels,
    cmap="coolwarm", center=0
)
plt.title("DIFF Heatmap (Edited - Baseline)\nPositive = more predictions, Negative = fewer")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

print("Done.")
```

```python
# =====================================================================
# GRID SEARCH FOR OPTIMAL DIRECTIONAL ERASURE
# =====================================================================

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Layer
from sklearn.metrics import confusion_matrix

TARGET_CLASS = C10_CLASS_TO_CORRUPT
LAYER_NAME = "conv5_block3_out"
NUM_SAMPLES = 300

ALPHA_VALUES = [0.25, 0.5, 0.75, 1.0, 1.25, 1.5]
PASSES = [1, 2, 3]
LAMBDA = 10.0     # penalize target accuracy

print("Running grid search for optimal directional erasure...")

# ---------------------------------------------------------
# Precompute activation direction
# ---------------------------------------------------------
activation_model = Model(
    inputs=trained_model.input,
    outputs=trained_model.get_layer(LAYER_NAME).output
)

x_class = x_train[np.argmax(y_train, axis=1) == TARGET_CLASS][:NUM_SAMPLES]
acts = activation_model.predict(x_class, batch_size=32)

v = acts.mean(axis=(0,1,2))
v = v / (np.linalg.norm(v) + 1e-8)

# ---------------------------------------------------------
# Direction eraser
# ---------------------------------------------------------
class DirectionEraser(Layer):
    def __init__(self, direction, alpha):
        super().__init__()
        self.v = tf.constant(direction.reshape(1,1,1,-1), dtype=tf.float32)
        self.alpha = alpha

    def call(self, h):
        dot = tf.reduce_sum(h * self.v, axis=-1, keepdims=True)
        return h - self.alpha * dot * self.v

# ---------------------------------------------------------
# Function to build edited model for given alpha + passes
# ---------------------------------------------------------
def build_erased_model(alpha, passes):
    model_up_to = Model(
        inputs=trained_model.input,
        outputs=trained_model.get_layer(LAYER_NAME).output
    )

    x = model_up_to.output
    for _ in range(passes):
        x = DirectionEraser(v, alpha=alpha)(x)

    layer_found = False
    for layer in trained_model.layers:
        if layer.name == LAYER_NAME:
            layer_found = True
            continue
        if layer_found:
            x = layer(x)

    return Model(inputs=trained_model.input, outputs=x)

# ---------------------------------------------------------
# Evaluate model
# ---------------------------------------------------------
results = []

y_true = np.argmax(y_test, axis=1)

for alpha in ALPHA_VALUES:
    for p in PASSES:
        print(f"Testing alpha={alpha}, passes={p}")

        edited = build_erased_model(alpha, p)
        y_pred = np.argmax(edited.predict(x_test, batch_size=64), axis=1)

        # Class-level stats
        cm = confusion_matrix(y_true, y_pred)
        class_acc = cm.diagonal() / cm.sum(axis=1)

        target_acc = class_acc[TARGET_CLASS]
        other_acc = np.mean(np.delete(class_acc, TARGET_CLASS))

        # scoring objective
        score = other_acc - LAMBDA * target_acc

        results.append({
            "alpha": alpha,
            "passes": p,
            "target_acc": target_acc,
            "other_acc": other_acc,
            "score": score,
            "model": edited
        })

# ---------------------------------------------------------
# Pick best setup
# ---------------------------------------------------------
best = sorted(results, key=lambda r: r["score"], reverse=True)[0]

print("\n===== BEST CONFIGURATION =====")
print("alpha:        ", best["alpha"])
print("passes:       ", best["passes"])
print("target_acc:   ", best["target_acc"])
print("other_acc:    ", best["other_acc"])
print("score:        ", best["score"])

best_model = best["model"]
```

```python
y_pred_best = np.argmax(best_model.predict(x_test, batch_size=64), axis=1)
cm_best = confusion_matrix(y_true, y_pred_best)
cm_best_norm = cm_best / cm_best.sum(axis=1, keepdims=True)

plt.figure(figsize=(10,8))
sns.heatmap(cm_best_norm, annot=True, fmt=".2f",
            xticklabels=cifar10_labels,
            yticklabels=cifar10_labels,
            cmap="magma")
plt.title("BEST Confusion Matrix After Grid-Search Directional Erasure")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```
