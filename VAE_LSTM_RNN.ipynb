```python
# ─────────────────────────────────────────────────────────────────────────────
# Single cell: build AE+LSTM model, load data, train, evaluate, and plot
# ─────────────────────────────────────────────────────────────────────────────
import os, pickle, numpy as np, matplotlib.pyplot as plt, seaborn as sns
from scipy.signal import resample
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras import Sequential, Model
from tensorflow.keras.layers import (
    Input, Flatten, Reshape, Dense, Dropout,
    Bidirectional, LSTM
)
from tensorflow.keras.optimizers import Adam

# --- 1) Build model builder with denoising autoencoder + LSTM stack ---
class LSTMModelBuilder:
    def __init__(self,
                 input_shape,
                 n_classes,
                 lr=1e-4,
                 lstm_units=128,
                 dropout_rates=(0.5,0.2,0.1),
                 bidirectional=False,
                 ae_enabled=False,
                 ae_hidden_dim=64,
                 ae_dropout=0.1):
        self.input_shape   = input_shape
        self.n_classes     = n_classes
        self.lr            = lr
        self.lstm_units    = lstm_units
        self.dr1, self.dr2, self.dr3 = dropout_rates
        self.bidirectional = bidirectional
        self.ae_enabled    = ae_enabled
        self.ae_hidden_dim = ae_hidden_dim
        self.ae_dropout    = ae_dropout

    def build(self):
        T, F = self.input_shape
        inp = Input(shape=(T, F), name="input_seq")
        x = inp

        # optional denoising autoencoder
        if self.ae_enabled:
            flat = Flatten()(x)
            enc  = Dense(self.ae_hidden_dim, activation="relu")(flat)
            enc  = Dropout(self.ae_dropout)(enc)
            dec  = Dense(T*F, activation="linear")(enc)
            dec  = Dropout(self.ae_dropout)(dec)
            x    = Reshape((T, F))(dec)

        # LSTM stack
        def LAYER(**kw):
            layer = LSTM(self.lstm_units, **kw)
            return Bidirectional(layer) if self.bidirectional else layer

        x = LAYER(input_shape=(T,F), return_sequences=True)(x)
        x = Dropout(self.dr1)(x)
        x = LAYER(return_sequences=False)(x)
        x = Dropout(self.dr2)(x)

        x = Dense(self.lstm_units, activation="relu")(x)
        x = Dropout(self.dr3)(x)
        out = Dense(self.n_classes, activation="softmax")(x)

        model = Model(inp, out)
        model.compile(
            optimizer=Adam(learning_rate=self.lr),
            loss="sparse_categorical_crossentropy",
            metrics=["accuracy"]
        )
        return model

    @staticmethod
    def upsample_sequences(X, target_len):
        return resample(X, target_len, axis=1)


# --- 2) Load & preprocess data ---
DATA_PATH = "RML2016.10a_dict.pkl"
with open(DATA_PATH, "rb") as f:
    raw = pickle.load(f, encoding="latin1")

X_list, y_list = [], []
for (mod, S), sigs in raw.items():
    if S < -6: continue
    for sig in sigs:
        iq   = np.vstack([sig[0], sig[1]]).T
        s_col= np.full((iq.shape[0],1), S, dtype=np.float32)
        X_list.append(np.hstack([iq, s_col]))
        y_list.append(mod)

X = np.array(X_list, dtype=np.float32)
y = np.array(y_list)
T, F = X.shape[1], X.shape[2]

# upsample to desired length if needed
desired_T = 128
if T != desired_T:
    X = LSTMModelBuilder.upsample_sequences(X, desired_T)

le     = LabelEncoder()
y_enc  = le.fit_transform(y)
n_cls  = len(le.classes_)

X_tr, X_te, y_tr, y_te = train_test_split(
    X, y_enc, test_size=0.2, stratify=y_enc, random_state=42
)

# --- 3) Instantiate builder & model ---
builder = LSTMModelBuilder(
    input_shape=(desired_T, F),
    n_classes=n_cls,
    lr=1e-4,
    lstm_units=256,
    dropout_rates=(0.3,0.3,0.2),
    bidirectional=True,
    ae_enabled=True,       # turn on AE
    ae_hidden_dim=128,
    ae_dropout=0.2
)
model = builder.build()
model.summary()

# --- 4) Train ---
history = model.fit(
    X_tr, y_tr,
    validation_data=(X_te, y_te),
    epochs=50,
    batch_size=64,
    verbose=2
)

# --- 5) Plot training curves ---
plt.figure(figsize=(6,4))
plt.plot(history.history["accuracy"], label="Train")
plt.plot(history.history["val_accuracy"], label="Val")
plt.xlabel("Epoch"); plt.ylabel("Accuracy")
plt.title("Train & Val Accuracy")
plt.legend(); plt.grid(True); plt.tight_layout()
plt.show()

# --- 6) Evaluate & confusion matrix ---
y_pred = model.predict(X_te).argmax(axis=1)
print(classification_report(y_te, y_pred, target_names=le.classes_))
cm = confusion_matrix(y_te, y_pred)
sns.heatmap(cm, annot=True, fmt="d", xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predicted"); plt.ylabel("True"); plt.title("Confusion Matrix")
plt.tight_layout(); plt.show()

# --- 7) Accuracy vs SNR ---
snr_vals = sorted({S for (_,S) in raw.keys()})
acc_snr  = []
for S in snr_vals:
    idx = [i for i,(mod,S2) in enumerate(raw.keys()) if S2==S]
    # rebuild per-SNR dataset
    X_s, y_s = [], []
    for (mod_val, s_val), sigs in raw.items():
        if s_val!=S: continue
        for sig in sigs:
            iq    = np.vstack([sig[0], sig[1]]).T
            s_col = np.full((iq.shape[0],1), s_val, dtype=np.float32)
            X_s.append(np.hstack([iq, s_col]))
            y_s.append(mod_val)
    if not X_s:
        acc_snr.append(np.nan)
        continue
    X_s = np.array(X_s, dtype=np.float32)
    if X_s.shape[1]!=desired_T:
        X_s = builder.upsample_sequences(X_s, desired_T)
    y_s_enc = le.transform(y_s)
    preds   = model.predict(X_s).argmax(axis=1)
    acc_snr.append((preds==y_s_enc).mean())

plt.figure(figsize=(7,4))
plt.plot(snr_vals, acc_snr, marker="o")
plt.xlabel("SNR [dB]"); plt.ylabel("Accuracy")
plt.title("Accuracy vs SNR")
plt.ylim(0,1.05); plt.grid(True); plt.tight_layout()
plt.show()
```

```python
import os, pickle, numpy as np, matplotlib.pyplot as plt, seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from tensorflow.keras import Input, Model
from tensorflow.keras.layers import (
    Conv1D, Conv1DTranspose, BatchNormalization, Activation,
    Add, GaussianNoise, Reshape, Flatten, Dense, Dropout
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import Sequence

# 1) Load & prepare data
with open("RML2016.10a_dict.pkl","rb") as f:
    raw = pickle.load(f, encoding="latin1")

X_list, y_list, snr_list = [], [], []
for (mod, snr), sigs in raw.items():
    for sig in sigs:
        iq = np.vstack([sig[0], sig[1]]).T
        X_list.append(np.hstack([iq, np.full((iq.shape[0],1), snr, dtype=np.float32)]))
        y_list.append(mod)
        snr_list.append(snr)

X_all   = np.array(X_list, dtype=np.float32)
y_all   = np.array(y_list)
snr_all = np.array(snr_list)

# dynamic clean/noisy split
clean_snr     = int(np.max(snr_all))
mask_clean    = (snr_all == clean_snr)
clean_classes = np.unique(y_all[mask_clean])
mask_noisy    = (snr_all >= -6) & (snr_all < clean_snr) & np.isin(y_all, clean_classes)

X_clean = X_all[mask_clean];  y_clean = y_all[mask_clean]
X_noisy = X_all[mask_noisy];  y_noisy = y_all[mask_noisy]

# encode labels
le = LabelEncoder(); le.fit(y_all)
y_noisy_enc = le.transform(y_noisy)
y_clean_enc = le.transform(y_clean)

# train/val split on noisy data
Xn_tr, Xn_va, yn_tr, yn_va = train_test_split(
    X_noisy, y_noisy_enc, test_size=0.2, stratify=y_noisy_enc, random_state=42
)

T, F = X_all.shape[1], X_all.shape[2]

# 2) Build 1D-UNet denoiser
def unet_denoiser(input_shape):
    inp = Input(shape=input_shape, name="unet_in")
    x = GaussianNoise(0.0)(inp)
    # Encoder
    c1 = Conv1D(64,5,padding="same")(x); c1=BatchNormalization()(c1); c1=Activation("relu")(c1)
    p1 = Conv1D(64,2,strides=2,padding="same")(c1)
    c2 = Conv1D(128,5,padding="same")(p1); c2=BatchNormalization()(c2); c2=Activation("relu")(c2)
    p2 = Conv1D(128,2,strides=2,padding="same")(c2)
    # Bottleneck
    b  = Conv1D(256,5,padding="same")(p2); b=BatchNormalization()(b); b=Activation("relu")(b)
    # Decoder
    u1 = Conv1DTranspose(128,2,strides=2,padding="same")(b); u1=Add()([u1,c2])
    c3 = Conv1D(128,5,padding="same")(u1); c3=BatchNormalization()(c3); c3=Activation("relu")(c3)
    u2 = Conv1DTranspose(64,2,strides=2,padding="same")(c3); u2=Add()([u2,c1])
    c4 = Conv1D(64,5,padding="same")(u2); c4=BatchNormalization()(c4); c4=Activation("relu")(c4)
    out=Conv1D(F,1,activation="linear",padding="same",name="unet_out")(c4)
    return Model(inp,out,name="unet_denoiser")

dae = unet_denoiser((T,F))
dae.summary()

# 3) Load & freeze classifier
clf = load_model("lstm_rnn_2024.keras")
for layer in clf.layers:
    layer.trainable = False

# 4) Joint model
noisy_in    = Input(shape=(T,F), name="noisy_in")
clean_recon = Input(shape=(T,F), name="clean_target")
denoised    = dae(noisy_in)
preds       = clf(denoised)

joint = Model([noisy_in,clean_recon],[denoised,preds])
joint.compile(
    optimizer=Adam(1e-4),
    loss=["mse","sparse_categorical_crossentropy"],
    loss_weights=[0.5,0.5],
    metrics=[None,"accuracy"]
)

# 5) Proper Sequence generator for noisy→clean pairing
class JointGen(Sequence):
    def __init__(self, Xn, yn, Xc, yc, bs=64):
        super(JointGen, self).__init__()  # ensure Sequence is initialized
        self.Xn, self.yn = Xn, yn
        self.Xc, self.yc = Xc, yc
        self.bs = bs
        self.clean_idx = {cls: np.where(self.yc==cls)[0] for cls in np.unique(self.yc)}
    def __len__(self):
        return max(1, len(self.Xn)//self.bs)
    def __getitem__(self, idx):
        i0 = idx*self.bs
        Xb = self.Xn[i0:i0+self.bs]
        yb = self.yn[i0:i0+self.bs]
        Xc_batch = np.zeros_like(Xb)
        for i, cls in enumerate(yb):
            ci = np.random.choice(self.clean_idx[cls])
            Xc_batch[i] = self.Xc[ci]
        return [Xb, Xc_batch], [Xc_batch, yb]

train_seq = JointGen(Xn_tr, yn_tr, X_clean, y_clean_enc, bs=64)
val_seq   = JointGen(Xn_va, yn_va, X_clean, y_clean_enc, bs=64)

# 6) EarlyStopping on classification accuracy
es = EarlyStopping(
    monitor="val_preds_accuracy", mode="max",
    min_delta=0.001, patience=5,
    restore_best_weights=True, verbose=1
)

# 7) Train
history = joint.fit(
    train_seq,
    validation_data=val_seq,
    epochs=50,
    callbacks=[es],
    verbose=2
)

# 8) Plot losses & validation accuracy
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history["loss"],label="total")
plt.plot(history.history["loss_1"],label="recon")
plt.plot(history.history["loss_2"],label="clf")
plt.legend(); plt.title("Losses"); plt.grid(True)
plt.subplot(1,2,2)
plt.plot(history.history["val_preds_accuracy"],label="val acc")
plt.axhline(0.9228,color="red",linestyle="--",label="baseline")
plt.legend(); plt.title("Val Accuracy"); plt.grid(True)
plt.tight_layout(); plt.show()

# 9) Final evaluation & confusion matrices
yb0 = clf.predict(Xn_va).argmax(axis=1)
yb1 = joint.predict([Xn_va, Xn_va])[1].argmax(axis=1)
acc0 = accuracy_score(yn_va, yb0)
acc1 = accuracy_score(yn_va, yb1)
print(f"Baseline: {acc0:.4f}, DAE+clf: {acc1:.4f}")

fig, axs = plt.subplots(1,2,figsize=(14,6))
for ax, pred, title in zip(axs, [yb0, yb1], ["Baseline", "Joint"]):
    cm = confusion_matrix(yn_va, pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=le.classes_, yticklabels=le.classes_, ax=ax)
    ax.set_title(title); ax.set_xlabel("Predicted"); ax.set_ylabel("True")
plt.tight_layout(); plt.show()

```

```python
# ─────────────────────────────────────────────────────────────────────────────
# Cell: Joint reconstruction + classification loss on a UNet denoiser + frozen LSTM
# ─────────────────────────────────────────────────────────────────────────────
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from tensorflow.keras import Input, Model
from tensorflow.keras.layers import (
    Conv1D, Conv1DTranspose, BatchNormalization, Activation,
    Add, GaussianNoise, Flatten, Dense, Dropout, Reshape
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import Sequence

# 1) Load & split data
with open("RML2016.10a_dict.pkl","rb") as f:
    raw = pickle.load(f, encoding="latin1")

X, y, snr = [], [], []
for (mod, s), sigs in raw.items():
    for sig in sigs:
        iq = np.vstack([sig[0], sig[1]]).T
        X.append(np.hstack([iq, np.full((iq.shape[0],1), s, dtype=np.float32)]))
        y.append(mod)
        snr.append(s)
X = np.array(X, dtype=np.float32); y = np.array(y); snr = np.array(snr)

# dynamic clean/noisy split
clean_snr     = int(snr.max())
mask_clean    = snr == clean_snr
clean_classes = np.unique(y[mask_clean])
mask_noisy    = (snr >= -6) & (snr < clean_snr) & np.isin(y, clean_classes)

X_clean = X[mask_clean]; y_clean = y[mask_clean]
X_noisy = X[mask_noisy]; y_noisy = y[mask_noisy]

# encode noisy labels
le           = LabelEncoder(); le.fit(y)
y_noisy_enc  = le.transform(y_noisy)

# train/val split on noisy
Xn_tr, Xn_va, yn_tr, yn_va = train_test_split(
    X_noisy, y_noisy_enc, test_size=0.2, stratify=y_noisy_enc, random_state=42
)

T, F = X.shape[1], X.shape[2]

# 2) UNet denoiser
def unet_denoiser(input_shape):
    inp = Input(shape=input_shape)
    x = GaussianNoise(0.0)(inp)
    # Encoder
    c1 = Conv1D(64,5,padding="same")(x); c1 = BatchNormalization()(c1); c1 = Activation("relu")(c1)
    p1 = Conv1D(64,2,strides=2,padding="same")(c1)
    c2 = Conv1D(128,5,padding="same")(p1); c2 = BatchNormalization()(c2); c2 = Activation("relu")(c2)
    p2 = Conv1D(128,2,strides=2,padding="same")(c2)
    # Bottleneck
    b  = Conv1D(256,5,padding="same")(p2); b = BatchNormalization()(b); b = Activation("relu")(b)
    # Decoder
    u1 = Conv1DTranspose(128,2,strides=2,padding="same")(b)
    u1 = Add()([u1, c2])
    c3 = Conv1D(128,5,padding="same")(u1); c3 = BatchNormalization()(c3); c3 = Activation("relu")(c3)
    u2 = Conv1DTranspose(64,2,strides=2,padding="same")(c3)
    u2 = Add()([u2, c1])
    c4 = Conv1D(64,5,padding="same")(u2); c4 = BatchNormalization()(c4); c4 = Activation("relu")(c4)
    out = Conv1D(F,1,activation="linear",padding="same")(c4)
    return Model(inp, out, name="unet_denoiser")

dae = unet_denoiser((T,F))

# 3) Load & freeze classifier
clf = load_model("lstm_rnn_2024.keras")
for layer in clf.layers: layer.trainable = False

# 4) Joint model: input noisy, two outputs (recon, class)
noisy_in   = Input(shape=(T,F), name="noisy_input")
clean_targ = Input(shape=(T,F), name="clean_target")
denoised   = dae(noisy_in)
preds      = clf(denoised)

joint = Model([noisy_in, clean_targ], [denoised, preds])
alpha = 0.5
joint.compile(
    optimizer=Adam(1e-4),
    loss=["mse","sparse_categorical_crossentropy"],
    loss_weights=[alpha,1-alpha],
    metrics=[None,"accuracy"]
)
joint.summary()

# 5) Generator yielding (noisy, clean) → (recon, label)
class JointGen(Sequence):
    def __init__(self, Xn, yn, Xc, yc, bs=64):
        self.Xn, self.yn = Xn, yn
        self.Xc, self.yc = Xc, yc
        self.bs = bs
        self.clean_idx = {cls: np.where(self.yc==cls)[0] for cls in np.unique(self.yc)}
    def __len__(self): return len(self.Xn)//self.bs
    def __getitem__(self, idx):
        i0 = idx*self.bs
        Xb = self.Xn[i0:i0+self.bs]
        yb = self.yn[i0:i0+self.bs]
        Xc_batch = np.zeros_like(Xb)
        for i, cls in enumerate(yb):
            ci = np.random.choice(self.clean_idx[cls])
            Xc_batch[i] = self.Xc[ci]
        return [Xb, Xc_batch], [Xc_batch, yb]

train_gen = JointGen(Xn_tr, yn_tr, X_clean, le.transform(y_clean), bs=64)
val_gen   = JointGen(Xn_va, yn_va, X_clean, le.transform(y_clean), bs=64)

# 6) Train with early stopping on classification accuracy
es = EarlyStopping(monitor="val_2_accuracy", min_delta=0.001, patience=5, restore_best_weights=True)
history = joint.fit(
    train_gen,
    validation_data=val_gen,
    epochs=50,
    callbacks=[es],
    verbose=2
)

# 7) Plot metrics
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history["loss"],    label="total")
plt.plot(history.history["loss_1"],  label="recon")
plt.plot(history.history["loss_2"],  label="clf")
plt.title("Losses"); plt.legend(); plt.grid(True)
plt.subplot(1,2,2)
plt.plot(history.history["val_2_accuracy"], label="val acc")
plt.axhline(0.9228, color="red", ls="--", label="baseline")
plt.title("Validation Acc"); plt.legend(); plt.grid(True)
plt.tight_layout(); plt.show()

# 8) Final evaluation
# Baseline
yb0 = clf.predict(Xn_va).argmax(axis=1)
acc0= accuracy_score(yn_va,yb0)
# Joint
yb1 = joint.predict([Xn_va, Xn_va])[1].argmax(axis=1)
acc1= accuracy_score(yn_va,yb1)
print(f"Baseline: {acc0:.4f}, Joint AE+clf: {acc1:.4f}")

# 9) Confusion matrices
fig,ax=plt.subplots(1,2,figsize=(14,6))
for i,(pred,title) in enumerate(zip([yb0,yb1],["Baseline","Joint"])):
    cm = confusion_matrix(yn_va, pred)
    sns.heatmap(cm, annot=True, fmt="d", ax=ax[i],
                xticklabels=le.classes_, yticklabels=le.classes_)
    ax[i].set_title(f"{title} Confusion"); ax[i].set_xlabel("Pred"); ax[i].set_ylabel("True")
plt.tight_layout(); plt.show()
```
