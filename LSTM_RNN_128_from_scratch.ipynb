```python
# ==============================================================
#  From‐scratch training of the original 128‐step LSTM‐RNN model
# ==============================================================
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# 1) Load the RML2016.10a dataset
with open("RML2016.10a_dict.pkl", "rb") as f:
    data = pickle.load(f, encoding="latin1")

# 2) Prepare samples: stack I,Q and append SNR column
X, y, snrs = [], [], []
for (mod, S), sigs in data.items():
    # keep only moderate-to-high SNR (e.g. > -6 dB)
    if S < -6:
        continue
    for sig in sigs:
        iq = np.vstack([sig[0], sig[1]]).T         # (128,2)
        s_col = np.full((128,1), S, dtype=np.float32)
        X.append(np.hstack([iq, s_col]))           # (128,3)
        y.append(mod)
X = np.array(X, dtype=np.float32)                # (N,128,3)
y = np.array(y)

print(f"Total samples: {X.shape[0]}  |  Sequence length: {X.shape[1]}")

# 3) Encode the labels
le = LabelEncoder()
y_enc = le.fit_transform(y)
n_classes = len(le.classes_)
print("Classes:", le.classes_)

# 4) Train/test split (stratified)
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y_enc, test_size=0.2, stratify=y_enc, random_state=42
)
print("Train/test split:", X_tr.shape[0], X_te.shape[0])

# 5) Build the LSTM‐RNN model
def build_model(input_shape, n_classes, lr=1e-4):
    m = Sequential([
        LSTM(128, input_shape=input_shape, return_sequences=True),
        Dropout(0.5),
        LSTM(128, return_sequences=False),
        Dropout(0.2),
        Dense(128, activation="relu"),
        Dropout(0.1),
        Dense(n_classes, activation="softmax")
    ])
    m.compile(
        optimizer=Adam(lr),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return m

model = build_model((128,3), n_classes)
model.summary()

# 6) Train from scratch
EPOCHS = 100
history = model.fit(
    X_tr, y_tr,
    validation_data=(X_te, y_te),
    epochs=EPOCHS, batch_size=64, verbose=2
)

# 7) Plot training & validation accuracy
plt.figure(figsize=(6,4))
plt.plot(history.history["accuracy"], label="Train")
plt.plot(history.history["val_accuracy"], label="Val")
plt.xlabel("Epoch"); plt.ylabel("Accuracy")
plt.title("Training & Validation Accuracy")
plt.legend(); plt.grid(True); plt.tight_layout()
plt.show()

# 8) Evaluate on test set
y_prob = model.predict(X_te, verbose=0)
y_pred = np.argmax(y_prob, axis=1)

print("\nClassification Report:")
print(classification_report(y_te, y_pred, target_names=le.classes_))

# 9) Confusion matrix
cm = confusion_matrix(y_te, y_pred)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predicted"); plt.ylabel("True")
plt.title("Confusion Matrix")
plt.tight_layout(); plt.show()

# 10) Accuracy vs SNR (-20 to +20 dB)
snr_values = list(range(-20, 22, 2))
acc_per_snr = []
for S in snr_values:
    # collect all test samples at exact SNR
    idx = [i for i,s in enumerate(data.keys()) if False]  # placeholder
    # instead rebuild on full data:
    idx = [i for i,(mod_val, s_val) in enumerate(data.keys()) if s_val == S]
    # simpler: re-create lists from data
    X_s, y_s = [], []
    for (mod_val, s_val), sigs in data.items():
        if s_val != S: continue
        for sig in sigs:
            iq = np.vstack([sig[0], sig[1]]).T
            s_col = np.full((128,1), s_val, dtype=np.float32)
            X_s.append(np.hstack([iq, s_col]))
            y_s.append(mod_val)
    if not X_s:
        acc_per_snr.append(np.nan)
        continue
    X_s = np.array(X_s, dtype=np.float32)
    y_s_enc = le.transform(y_s)
    preds = model.predict(X_s, verbose=0).argmax(axis=1)
    acc_per_snr.append((preds == y_s_enc).mean())

plt.figure(figsize=(7,4))
plt.plot(snr_values, acc_per_snr, marker="o")
plt.xlabel("SNR [dB]"); plt.ylabel("Accuracy")
plt.title("Accuracy vs SNR")
plt.ylim(0,1.05); plt.grid(True); plt.tight_layout()
plt.show()
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# Cell: Train with early stopping when val_accuracy stops improving by >0.001 for 5 epochs
# ─────────────────────────────────────────────────────────────────────────────
from tensorflow.keras.callbacks import EarlyStopping

# Configure early stopping: stop if validation accuracy doesn't improve by at least 0.001 for 5 epochs
early_stop = EarlyStopping(
    monitor='val_accuracy',
    min_delta=0.001,
    patience=5,
    restore_best_weights=True,
    verbose=1
)

# Train (set a high max epoch count; actual stop will be triggered by EarlyStopping)
max_epochs = 1000
history_es = model.fit(
    X_tr, y_tr,
    validation_data=(X_te, y_te),
    epochs=max_epochs,
    batch_size=64,
    callbacks=[early_stop],
    verbose=2
)

# Plot training & validation accuracy to see where it stopped
plt.figure(figsize=(6,4))
plt.plot(history_es.history["accuracy"], label="Train")
plt.plot(history_es.history["val_accuracy"], label="Val")
plt.xlabel("Epoch"); plt.ylabel("Accuracy")
plt.title("Training & Validation Accuracy with Early Stopping")
plt.legend(); plt.grid(True); plt.tight_layout()
plt.show()
```
