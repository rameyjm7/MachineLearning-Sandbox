```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import load_model
from tensorflow.keras.callbacks import EarlyStopping

from scipy.signal import resample
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    classification_report, confusion_matrix, ConfusionMatrixDisplay
)
from tensorflow.keras import Sequential, Model
from tensorflow.keras.layers import (
    Bidirectional, LSTM, Dropout, Dense, Input
)
from tensorflow.keras.optimizers import Adam

# ============================================================================
# 1) MODEL BUILDER
# ============================================================================
class LSTMModelBuilder:
    def __init__(self,
                 input_shape,
                 n_classes,
                 lr=1e-4,
                 lstm_units=128,
                 dropout_rates=(0.5, 0.2, 0.1),
                 bidirectional=False):
        """
        input_shape: tuple, e.g. (timesteps, features)
        n_classes: int
        lr: learning rate
        lstm_units: hidden size of each LSTM
        dropout_rates: tuple of three floats for
                       (after 1st LSTM, after 2nd LSTM, after Dense)
        bidirectional: wrap LSTMs in Bidirectional if True
        """
        self.input_shape   = input_shape
        self.n_classes     = n_classes
        self.lr            = lr
        self.lstm_units    = lstm_units
        self.dr1, self.dr2, self.dr3 = dropout_rates
        self.bidirectional = bidirectional

    def build(self):
        m = Sequential()
        LAYER = lambda **kw: (
            Bidirectional(LSTM(self.lstm_units, **kw))
            if self.bidirectional
            else LSTM(self.lstm_units, **kw)
        )

        # first LSTM (keep sequences)
        m.add(LAYER(input_shape=self.input_shape, return_sequences=True))
        m.add(Dropout(self.dr1))

        # second LSTM (no sequences out)
        m.add(LAYER(return_sequences=False))
        m.add(Dropout(self.dr2))

        # dense head
        m.add(Dense(self.lstm_units, activation="relu"))
        m.add(Dropout(self.dr3))

        # final softmax
        m.add(Dense(self.n_classes, activation="softmax"))

        m.compile(
            optimizer=Adam(learning_rate=self.lr),
            loss="sparse_categorical_crossentropy",
            metrics=["accuracy"]
        )
        return m

    @staticmethod
    def upsample_sequences(X, target_len):
        """Resample axis-1 to `target_len` timesteps."""
        return resample(X, target_len, axis=1)


# ============================================================================
# 2) TRAIN / EVAL / PLOTTING PIPELINE
# ============================================================================
class SignalClassifierTrainer:
    def __init__(self, model_builder, snr_threshold=-6,
                 test_size=0.2, random_state=42):
        """
        model_builder: instance of LSTMModelBuilder
        snr_threshold: drop samples below this SNR
        test_size, random_state: for train/test split
        """
        self.mb           = model_builder
        self.snr_thresh   = snr_threshold
        self.test_size    = test_size
        self.random_state = random_state

        # Will be filled in load_data()
        self.raw_data = None
        self.X, self.y = None, None
        self.le       = None
        self.X_tr, self.X_te, self.y_tr, self.y_te = (None,)*4
        self.model    = None
        self.history  = None

    def load_data(self, path):
        """Loads RML2016.10a_dict.pkl and builds X,y arrays."""
        with open(path, "rb") as f:
            data = pickle.load(f, encoding="latin1")
        self.raw_data = data

        X_list, y_list = [], []
        for (mod, S), sigs in data.items():
            if S < self.snr_thresh:
                continue
            for sig in sigs:
                iq   = np.vstack([sig[0], sig[1]]).T  # (T,2)
                s_col= np.full((iq.shape[0],1), S, dtype=np.float32)
                X_list.append(np.hstack([iq, s_col])) # (T,3)
                y_list.append(mod)

        X = np.array(X_list, dtype=np.float32)
        y = np.array(y_list)

        # optionally upsample if builder.input_shape[0] != X.shape[1]
        T_desired = self.mb.input_shape[0]
        if X.shape[1] != T_desired:
            X = self.mb.upsample_sequences(X, T_desired)

        # encode
        le     = LabelEncoder()
        y_enc  = le.fit_transform(y)

        self.X, self.y = X, y_enc
        self.le       = le

    def split_data(self):
        """Stratified train/test split."""
        self.X_tr, self.X_te, self.y_tr, self.y_te = train_test_split(
            self.X, self.y,
            test_size=self.test_size,
            stratify=self.y,
            random_state=self.random_state
        )

    def build_model(self):
        """Use the builder to get a compiled model."""
        self.model = self.mb.build()

    def train(self, epochs=100, batch_size=64, verbose=2):
        """Fit and store history."""
        self.history = self.model.fit(
            self.X_tr, self.y_tr,
            validation_data=(self.X_te, self.y_te),
            epochs=epochs,
            batch_size=batch_size,
            verbose=verbose
        )

    def plot_training_history(self):
        """Plot accuracy curves."""
        plt.figure(figsize=(6,4))
        plt.plot(self.history.history["accuracy"], label="Train")
        plt.plot(self.history.history["val_accuracy"], label="Val")
        plt.xlabel("Epoch"); plt.ylabel("Accuracy")
        plt.title("Training & Validation Accuracy")
        plt.legend(); plt.grid(True); plt.tight_layout()
        plt.show()

    def evaluate(self):
        """Print classification report on test set."""
        y_prob = self.model.predict(self.X_te, verbose=0)
        y_pred = np.argmax(y_prob, axis=1)
        print("\nClassification Report:")
        print(classification_report(
            self.y_te, y_pred, target_names=self.le.classes_
        ))
        return y_pred

    def plot_confusion_matrix(self):
        """Show a heatmap of the confusion matrix."""
        y_pred = np.argmax(self.model.predict(self.X_te, verbose=0), axis=1)
        cm = confusion_matrix(self.y_te, y_pred)
        plt.figure(figsize=(10,8))
        sns.heatmap(
            cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=self.le.classes_,
            yticklabels=self.le.classes_
        )
        plt.xlabel("Predicted"); plt.ylabel("True")
        plt.title("Confusion Matrix")
        plt.tight_layout(); plt.show()

    def plot_accuracy_vs_snr(self, snr_values):
        """Rebuild per-SNR accuracy curve."""
        acc_per_snr = []
        for S in snr_values:
            X_s, y_s = [], []
            for (mod_val, s_val), sigs in self.raw_data.items():
                if s_val != S:
                    continue
                for sig in sigs:
                    iq    = np.vstack([sig[0], sig[1]]).T
                    s_col = np.full((iq.shape[0],1), s_val, dtype=np.float32)
                    X_s.append(np.hstack([iq, s_col]))
                    y_s.append(mod_val)

            if not X_s:
                acc_per_snr.append(np.nan)
                continue

            X_s = np.array(X_s, dtype=np.float32)
            # upsample if needed
            if X_s.shape[1] != self.mb.input_shape[0]:
                X_s = self.mb.upsample_sequences(
                    X_s, self.mb.input_shape[0]
                )
            y_s_enc = self.le.transform(y_s)
            preds = self.model.predict(X_s, verbose=0).argmax(axis=1)
            acc_per_snr.append((preds == y_s_enc).mean())

        plt.figure(figsize=(7,4))
        plt.plot(snr_values, acc_per_snr, marker="o")
        plt.xlabel("SNR [dB]"); plt.ylabel("Accuracy")
        plt.title("Accuracy vs SNR")
        plt.ylim(0,1.05); plt.grid(True); plt.tight_layout()
        plt.show()


# ============================================================================
# 3) USAGE
# ============================================================================

if __name__ == "__main__":
    DATA_PATH = "RML2016.10a_dict.pkl"

    # 1) instantiate builder (use original 128-step, 128 units, bidirectional)
    builder = LSTMModelBuilder(
        input_shape=(128,3),
        n_classes=None,       # we'll fill this after loading data
        lr=1e-4,
        lstm_units=128,
        dropout_rates=(0.5,0.2,0.1),
        bidirectional=False
    )

    # 2) trainer
    trainer = SignalClassifierTrainer(builder, snr_threshold=-6)

    # 3) load & encode
    trainer.load_data(DATA_PATH)
    builder.n_classes = trainer.le.classes_.shape[0]  # now we know n_classes

    # 4) split
    trainer.split_data()

    # 5) build, train
    trainer.build_model()
    trainer.train(epochs=100, batch_size=64)

    # 6) plots & evaluation
    trainer.plot_training_history()
    trainer.evaluate()
    trainer.plot_confusion_matrix()
    trainer.plot_accuracy_vs_snr(list(range(-20,22,2)))


    
    # adjust the path if needed
    model_path = "saved_models/signal_classifier_final.keras"
    model = load_model(model_path)
    
    # if you want to hook it back into the trainer:
    trainer.model = model
    
    
    # ─────────────────────────────────────────────────────────────────────────────
    # Cell: Continue training with early stopping instead of fixed 300 epochs
    # ─────────────────────────────────────────────────────────────────────────────
    
    # EarlyStopping: halt if val_accuracy doesn’t improve by >0.001 for 5 epochs
    early_stop = EarlyStopping(
        monitor='val_accuracy',
        min_delta=0.001,
        patience=5,
        restore_best_weights=True,
        verbose=1
    )
    
    # Continue fitting the same model, up to 300 more epochs but stopping early
    history2 = trainer.model.fit(
        trainer.X_tr, trainer.y_tr,
        validation_data=(trainer.X_te, trainer.y_te),
        epochs=300,
        batch_size=64*2,
        callbacks=[early_stop],
        verbose=2
    )
    
    # Merge the new history into the original
    for metric, new_vals in history2.history.items():
        if metric in trainer.history.history:
            trainer.history.history[metric].extend(new_vals)
        else:
            trainer.history.history[metric] = new_vals
    
    # Re-plot combined training curves
    trainer.plot_training_history()
    
    # Re-evaluate & re-plot confusion matrix
    trainer.evaluate()
    trainer.plot_confusion_matrix()
    
    # Update Accuracy vs SNR curve
    trainer.plot_accuracy_vs_snr(list(range(-20, 22, 2)))
```

```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Bidirectional, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from scipy.signal import resample

# ─────────────────────────────────────────────────────────────────────────────
# LSTM MODEL BUILDER
# ─────────────────────────────────────────────────────────────────────────────
class LSTMModelBuilder:
    def __init__(self, input_shape, n_classes, lr=1e-4, lstm_units=128,
                 dropout_rates=(0.5, 0.2, 0.1), bidirectional=False):
        self.input_shape = input_shape
        self.n_classes = n_classes
        self.lr = lr
        self.lstm_units = lstm_units
        self.dr1, self.dr2, self.dr3 = dropout_rates
        self.bidirectional = bidirectional

    def build(self):
        model = Sequential()
        layer = lambda **kw: (
            Bidirectional(LSTM(self.lstm_units, **kw)) if self.bidirectional
            else LSTM(self.lstm_units, **kw)
        )
        model.add(layer(input_shape=self.input_shape, return_sequences=True))
        model.add(Dropout(self.dr1))
        model.add(layer(return_sequences=False))
        model.add(Dropout(self.dr2))
        model.add(Dense(self.lstm_units, activation="relu"))
        model.add(Dropout(self.dr3))
        model.add(Dense(self.n_classes, activation="softmax"))
        model.compile(optimizer=Adam(learning_rate=self.lr),
                      loss="sparse_categorical_crossentropy",
                      metrics=["accuracy"])
        return model


# ─────────────────────────────────────────────────────────────────────────────
# DATA LOADING WITH 512 SAMPLES + ΔPHASE + SNR
# ─────────────────────────────────────────────────────────────────────────────
def load_data_512_with_dphase(path, snr_thresh=-6):
    with open(path, "rb") as f:
        raw = pickle.load(f, encoding="latin1")
    X_list, y_list = [], []
    for (mod, S), sigs in raw.items():
        if S < snr_thresh:
            continue
        for i in range(0, len(sigs) - 3, 4):
            sigs_concat = np.vstack([
                np.vstack([sigs[i+k][0], sigs[i+k][1]]).T for k in range(4)
            ])  # (512,2)
            phase = np.angle(sigs_concat[:, 0] + 1j * sigs_concat[:, 1])
            dphase = np.diff(np.unwrap(phase), prepend=phase[0]).reshape(-1, 1)
            snr_col = np.full((512, 1), S, dtype=np.float32)
            features = np.hstack([sigs_concat, snr_col, dphase])  # (512, 4)
            X_list.append(features)
            y_list.append(mod)
    X = np.array(X_list, dtype=np.float32)
    y = np.array(y_list)
    return X, y, raw


# ─────────────────────────────────────────────────────────────────────────────
# PLOTTING FUNCTIONS
# ─────────────────────────────────────────────────────────────────────────────
def plot_history(history):
    plt.figure(figsize=(6, 4))
    plt.plot(history.history['accuracy'], label="Train")
    plt.plot(history.history['val_accuracy'], label="Val")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Training & Validation Accuracy")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()


def plot_confusion(y_true, y_pred, label_encoder):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=label_encoder.classes_,
                yticklabels=label_encoder.classes_)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title("Confusion Matrix")
    plt.tight_layout()
    plt.show()


def plot_accuracy_vs_snr(model, raw_data, le, input_shape):
    snrs = list(range(-20, 22, 2))
    accs = []
    for S in snrs:
        X_s, y_s = [], []
        for (mod, s_val), sigs in raw_data.items():
            if s_val != S:
                continue
            for i in range(0, len(sigs) - 3, 4):
                sigs_concat = np.vstack([
                    np.vstack([sigs[i+k][0], sigs[i+k][1]]).T for k in range(4)
                ])
                phase = np.angle(sigs_concat[:, 0] + 1j * sigs_concat[:, 1])
                dphase = np.diff(np.unwrap(phase), prepend=phase[0]).reshape(-1, 1)
                snr_col = np.full((512, 1), S, dtype=np.float32)
                features = np.hstack([sigs_concat, snr_col, dphase])
                X_s.append(features)
                y_s.append(mod)
        if not X_s:
            accs.append(np.nan)
            continue
        X_s = np.array(X_s, dtype=np.float32)
        y_s_enc = le.transform(y_s)
        preds = model.predict(X_s, verbose=0).argmax(axis=1)
        accs.append((preds == y_s_enc).mean())

    plt.figure(figsize=(7, 4))
    plt.plot(snrs, accs, marker="o")
    plt.xlabel("SNR [dB]")
    plt.ylabel("Accuracy")
    plt.title("Accuracy vs SNR")
    plt.ylim(0, 1.05)
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# MAIN PIPELINE
# ─────────────────────────────────────────────────────────────────────────────
DATA_PATH = "RML2016.10a_dict.pkl"
X, y, raw_data = load_data_512_with_dphase(DATA_PATH, snr_thresh=-6)

le = LabelEncoder()
y_enc = le.fit_transform(y)

X_tr, X_te, y_tr, y_te = train_test_split(
    X, y_enc, test_size=0.2, stratify=y_enc, random_state=42
)

builder = LSTMModelBuilder(input_shape=(512, 4), n_classes=len(le.classes_))
model = builder.build()

early_stop = EarlyStopping(
    monitor="val_accuracy",
    min_delta=0.001,
    patience=7,
    restore_best_weights=True,
    verbose=1
)

history = model.fit(
    X_tr, y_tr,
    validation_data=(X_te, y_te),
    epochs=100,
    batch_size=64,
    callbacks=[early_stop],
    verbose=2
)

plot_history(history)

y_pred = model.predict(X_te, verbose=0).argmax(axis=1)
print("\nClassification Report:")
print(classification_report(y_te, y_pred, target_names=le.classes_))

plot_confusion(y_te, y_pred, le)
plot_accuracy_vs_snr(model, raw_data, le, input_shape=(512, 4))
```

```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Bidirectional, Dropout, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
import seaborn as sns

# ─────────────────────────────────────────────────────────────────────────────
# SETTINGS
# ─────────────────────────────────────────────────────────────────────────────
DATA_PATH = "RML2016.10a_dict.pkl"
INPUT_SHAPE = (512, 4)
SNR_THRESH = -6
INITIAL_CLASSES = list(range(0, 5))  # Classes 0 to 4
INCREMENTAL_CLASSES = list(range(5, 11))  # Add one at a time
SOFTMAX_THRESHOLD = 0.7


# ─────────────────────────────────────────────────────────────────────────────
# HELPERS
# ─────────────────────────────────────────────────────────────────────────────
def extract_features(raw_data, allowed_labels, le):
    X_list, y_list = [], []
    for (label, snr), sigs in raw_data.items():
        if snr < SNR_THRESH or label not in allowed_labels:
            continue
        for i in range(0, len(sigs) - 3, 4):
            segs = np.vstack([np.vstack([sigs[i+k][0], sigs[i+k][1]]).T for k in range(4)])  # (512,2)
            phase = np.angle(segs[:, 0] + 1j * segs[:, 1])
            dphase = np.diff(np.unwrap(phase), prepend=phase[0]).reshape(-1, 1)
            snr_col = np.full((512, 1), snr, dtype=np.float32)
            X_list.append(np.hstack([segs, snr_col, dphase]))
            y_list.append(label)
    X = np.array(X_list, dtype=np.float32)
    y = np.array(y_list)
    return X, le.transform(y)


def build_model(n_classes, input_shape=INPUT_SHAPE, lstm_units=128, dropout=(0.5, 0.2, 0.1)):
    m = Sequential()
    m.add(LSTM(lstm_units, return_sequences=True, input_shape=input_shape))
    m.add(Dropout(dropout[0]))
    m.add(LSTM(lstm_units, return_sequences=False))
    m.add(Dropout(dropout[1]))
    m.add(Dense(lstm_units, activation="relu"))
    m.add(Dropout(dropout[2]))
    m.add(Dense(n_classes, activation="softmax"))
    m.compile(optimizer=Adam(1e-4), loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return m


def evaluate_model(model, X_test, y_test, le, title=""):
    y_pred = model.predict(X_test, verbose=0).argmax(axis=1)

    # Extract only the label names present in y_test
    used_labels = sorted(set(y_test))
    label_names = le.inverse_transform(used_labels)

    print("\n" + title)
    print(classification_report(y_test, y_pred, labels=used_labels, target_names=label_names))

    cm = confusion_matrix(y_test, y_pred, labels=used_labels)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=label_names,
                yticklabels=label_names)
    plt.title(f"Confusion Matrix - {title}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()


def expand_model(old_model, new_n_classes):
    # Extract existing weights
    old_weights = old_model.get_weights()
    old_n_classes = old_model.output_shape[-1]

    # Build new model
    new_model = build_model(new_n_classes)

    # Transfer weights (excluding final Dense layer)
    for i in range(len(new_model.layers) - 1):
        new_model.layers[i].set_weights(old_model.layers[i].get_weights())

    # Final Dense layer: expand weights and biases
    old_W, old_b = old_model.layers[-1].get_weights()
    new_W = np.zeros((old_W.shape[0], new_n_classes), dtype=np.float32)
    new_b = np.zeros((new_n_classes,), dtype=np.float32)
    new_W[:, :old_n_classes] = old_W
    new_b[:old_n_classes] = old_b
    new_model.layers[-1].set_weights([new_W, new_b])

    return new_model


# ─────────────────────────────────────────────────────────────────────────────
# MAIN LOOP
# ─────────────────────────────────────────────────────────────────────────────
# Load data and encode
with open(DATA_PATH, "rb") as f:
    raw = pickle.load(f, encoding="latin1")
modulations = sorted(set(k[0] for k in raw.keys()))
le = LabelEncoder()
le.fit(modulations)
print("Label mapping:")
for idx, name in enumerate(le.classes_):
    print(f"{idx:2} → {name}")

# Train on classes 0–4
X_train, y_train = extract_features(raw, [le.classes_[i] for i in INITIAL_CLASSES], le)
X_tr, X_te, y_tr, y_te = train_test_split(X_train, y_train, test_size=0.2, stratify=y_train, random_state=42)

model = build_model(n_classes=len(INITIAL_CLASSES))
print(f"\nTraining on initial classes {INITIAL_CLASSES}...")
model.fit(X_tr, y_tr, validation_data=(X_te, y_te), epochs=25, batch_size=64, verbose=2)

evaluate_model(model, X_te, y_te, le, title="Initial Model (0–4)")

# Incremental learning: classes 5 → 10
for new_class in INCREMENTAL_CLASSES:
    new_label = le.classes_[new_class]
    print(f"\n🔁 Incrementally adding class {new_class}: {new_label}")

    # Extract only the new class
    X_new, y_new = extract_features(raw, [new_label], le)
    y_new[:] = new_class  # force relabeling (already correct, just for clarity)

    # Expand the model
    model = expand_model(model, new_n_classes=new_class + 1)

    # Combine old + new data
    X_tr_combined = np.vstack([X_tr, X_new])
    y_tr_combined = np.hstack([y_tr, y_new])

    # Retrain on expanded dataset
    X_tr_final, X_val_final, y_tr_final, y_val_final = train_test_split(
        X_tr_combined, y_tr_combined,
        test_size=0.2, stratify=y_tr_combined, random_state=42
    )
    model.fit(X_tr_final, y_tr_final,
              validation_data=(X_val_final, y_val_final),
              epochs=10, batch_size=64, verbose=2)

    # Evaluate
    evaluate_model(model, X_te, y_te, le, title=f"After Adding {new_label}")
```

```python
import numpy as np, pickle, matplotlib.pyplot as plt, seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import classification_report, confusion_matrix, silhouette_score
import tensorflow as tf
from tensorflow.keras import Input, Model
from tensorflow.keras.layers import (
    LSTM, Dense, Dropout, Concatenate, GlobalAveragePooling2D
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import ResNet50

# ─────────────────────────────────────────────────────────────────────────────
# ⚙️ DATA HELPERS
# ─────────────────────────────────────────────────────────────────────────────
def extract_features(raw, allowed_labels, le, snr_thresh=-6):
    X_seq, X_fft, y = [], [], []
    for (label, snr), sigs in raw.items():
        if snr < snr_thresh or label not in allowed_labels: continue
        for i in range(0, len(sigs) - 3, 4):
            iq = np.vstack([np.vstack([sigs[i+k][0], sigs[i+k][1]]).T for k in range(4)])  # (512,2)
            phase = np.angle(iq[:,0] + 1j*iq[:,1])
            dphase = np.diff(np.unwrap(phase), prepend=phase[0]).reshape(-1,1)
            snr_col = np.full((512,1), snr, dtype=np.float32)
            X_seq.append(np.hstack([iq, snr_col, dphase]))

            fft = np.fft.fftshift(np.fft.fft(iq[:,0] + 1j*iq[:,1], n=1024))
            fft_mag = np.log1p(np.abs(fft[:1024]))
            fft_img = fft_mag[:1024].reshape(32, 32)
            X_fft.append(np.stack([fft_img]*3, axis=-1))  # (32,32,3)

            y.append(label)
    return np.array(X_seq, dtype=np.float32), np.array(X_fft, dtype=np.float32), le.transform(y)

def detect_unknowns(model, X_seq, X_fft, threshold=0.7):
    probs = model.predict([X_seq, X_fft], verbose=0)
    return np.where(probs.max(axis=1) < threshold)[0]

def cluster_novel(X, max_clusters=4):
    flat = X.reshape(len(X), -1)
    best_score = -1
    for k in range(2, min(max_clusters, len(X)) + 1):
        km = KMeans(n_clusters=k).fit(flat)
        score = silhouette_score(flat, km.labels_)
        if score > best_score:
            best_km = km
            best_score = score
    print(f"🧠 Clustering result: {best_km.n_clusters} clusters, silhouette={best_score:.2f}")
    return best_km.labels_

# ─────────────────────────────────────────────────────────────────────────────
# ⚙️ MODEL: FUSED LSTM + RESNET BRANCHES
# ─────────────────────────────────────────────────────────────────────────────
def build_fused_model(n_classes):
    # LSTM branch
    inp_seq = Input(shape=(512, 4), name="iq_seq")
    x = LSTM(128, return_sequences=True)(inp_seq)
    x = Dropout(0.5)(x)
    x = LSTM(128)(x)
    x = Dropout(0.2)(x)
    x = Dense(128, activation="relu")(x)
    x = Dropout(0.1)(x)

    # ResNet branch
    inp_fft = Input(shape=(32, 32, 3), name="fft_img")
    base = ResNet50(weights="imagenet", include_top=False, input_shape=(32,32,3))
    base.trainable = False
    y = base(inp_fft)
    y = GlobalAveragePooling2D()(y)
    y = Dense(256, activation="relu")(y)
    y = Dropout(0.3)(y)

    # Fusion head
    fused = Concatenate()([x, y])
    fused = Dense(128, activation="relu")(fused)
    out = Dense(n_classes, activation="softmax")(fused)

    model = Model(inputs=[inp_seq, inp_fft], outputs=out)
    model.compile(optimizer=Adam(1e-4), loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return model

def expand_fused_model(old_model, new_n_classes):
    new_model = build_fused_model(new_n_classes)
    for i in range(len(new_model.layers) - 1):
        if i < len(old_model.layers):
            try:
                new_model.layers[i].set_weights(old_model.layers[i].get_weights())
            except:
                pass
    old_W, old_b = old_model.layers[-1].get_weights()
    new_W = np.zeros((old_W.shape[0], new_n_classes), dtype=np.float32)
    new_b = np.zeros((new_n_classes,), dtype=np.float32)
    new_W[:, :old_W.shape[1]] = old_W
    new_b[:old_b.shape[0]] = old_b
    new_model.layers[-1].set_weights([new_W, new_b])
    return new_model

def evaluate(model, X_seq, X_fft, y, le, title=""):
    y_pred = model.predict([X_seq, X_fft], verbose=0).argmax(axis=1)
    labels_used = sorted(set(y))
    names = le.inverse_transform(labels_used)
    print(f"\n📊 {title}")
    print(classification_report(y, y_pred, labels=labels_used, target_names=names))
    cm = confusion_matrix(y, y_pred, labels=labels_used)
    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt="d", xticklabels=names, yticklabels=names, cmap="Blues")
    plt.xlabel("Predicted"); plt.ylabel("True"); plt.title(title)
    plt.tight_layout(); plt.show()

# ─────────────────────────────────────────────────────────────────────────────
# 🚀 OSIL LOOP
# ─────────────────────────────────────────────────────────────────────────────
with open("RML2016.10a_dict.pkl", "rb") as f:
    raw = pickle.load(f, encoding="latin1")

le = LabelEncoder()
le.fit(sorted(set(k[0] for k in raw)))
print("Modulation map:")
for i, name in enumerate(le.classes_): print(f"{i:2} → {name}")

# Phase 1: train on classes 0–4
initial_classes = list(le.classes_[:5])
X_seq, X_fft, y = extract_features(raw, initial_classes, le)
X_seq_tr, X_seq_te, X_fft_tr, X_fft_te, y_tr, y_te = train_test_split(X_seq, X_fft, y, test_size=0.2, stratify=y)

model = build_fused_model(n_classes=5)
model.fit([X_seq_tr, X_fft_tr], y_tr, validation_data=([X_seq_te, X_fft_te], y_te), epochs=15, batch_size=64, verbose=2)
evaluate(model, X_seq_te, X_fft_te, y_te, le, "Initial Training (0–4)")

# Phase 2: Incrementally add classes 5 → 10
for new_id in range(5, 11):
    label_name = le.classes_[new_id]
    print(f"\n🔍 Detecting novel class: {label_name}")
    X_seq_new, X_fft_new, y_new_full = extract_features(raw, [label_name], le)
    unknown_idx = detect_unknowns(model, X_seq_new, X_fft_new, threshold=0.7)
    if len(unknown_idx) == 0:
        print("⚠️ No unknowns detected!")
        continue

    X_seq_unknown = X_seq_new[unknown_idx]
    X_fft_unknown = X_fft_new[unknown_idx]

    cluster_labels = cluster_novel(X_seq_unknown, max_clusters=1)
    for cl in np.unique(cluster_labels):
        X_seq_cl = X_seq_unknown[cluster_labels == cl]
        X_fft_cl = X_fft_unknown[cluster_labels == cl]
        y_cl = np.full(len(X_seq_cl), new_id)

        print(f"➕ Adding class {new_id}: {label_name} with {len(X_seq_cl)} samples")
        model = expand_fused_model(model, new_n_classes=new_id + 1)
        X_seq_tr = np.vstack([X_seq_tr, X_seq_cl])
        X_fft_tr = np.vstack([X_fft_tr, X_fft_cl])
        y_tr = np.hstack([y_tr, y_cl])

        # Retrain
        Xs, Xv, Fs, Fv, ys, yv = train_test_split(X_seq_tr, X_fft_tr, y_tr, stratify=y_tr, test_size=0.2)
        model.fit([Xs, Fs], ys, validation_data=([Xv, Fv], yv), epochs=8, batch_size=64, verbose=2)
        evaluate(model, X_seq_te, X_fft_te, y_te, le, f"Model after class {label_name}")
```
