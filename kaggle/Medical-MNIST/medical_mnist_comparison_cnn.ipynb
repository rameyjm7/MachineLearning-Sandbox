```python
import os
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import confusion_matrix, classification_report
import kagglehub

# ----------------------------
# Dataset setup
# ----------------------------
path = kagglehub.dataset_download("andrewmvd/medical-mnist")
dataset_dir = path
img_size = (64, 64)
batch_size = 32
base_lr = 1e-4
epochs = 8

datagen = ImageDataGenerator(rescale=1.0/255, validation_split=0.2)

train_gen = datagen.flow_from_directory(
    dataset_dir,
    target_size=img_size,
    color_mode="rgb",
    batch_size=batch_size,
    class_mode="categorical",
    subset="training",
    shuffle=True
)

val_gen = datagen.flow_from_directory(
    dataset_dir,
    target_size=img_size,
    color_mode="rgb",
    batch_size=batch_size,
    class_mode="categorical",
    subset="validation",
    shuffle=False
)

num_classes = len(train_gen.class_indices)
class_labels = list(train_gen.class_indices.keys())

# ----------------------------
# Model definitions
# ----------------------------

def build_tinycnn(input_shape, num_classes):
    model = Sequential([
        Conv2D(16, (3,3), activation='relu', input_shape=input_shape),
        MaxPooling2D(2,2),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer=Adam(base_lr), loss="categorical_crossentropy", metrics=["accuracy"])
    return model


def build_reducedfilter_cnn(input_shape, num_classes):
    model = Sequential([
        Conv2D(8, (3,3), activation='relu', input_shape=input_shape),
        MaxPooling2D(2,2),
        Flatten(),
        Dense(32, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer=Adam(base_lr), loss="categorical_crossentropy", metrics=["accuracy"])
    return model


def build_globalavgpool_cnn(input_shape, num_classes):
    model = Sequential([
        Conv2D(8, (3,3), activation='relu', input_shape=input_shape),
        MaxPooling2D(2,2),
        Conv2D(8, (3,3), activation='relu'),
        GlobalAveragePooling2D(),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer=Adam(base_lr), loss="categorical_crossentropy", metrics=["accuracy"])
    return model

# ----------------------------
# Training loop
# ----------------------------
models = {
    "TinyCNN": build_tinycnn((*img_size, 3), num_classes),
    "ReducedFilterCNN": build_reducedfilter_cnn((*img_size, 3), num_classes),
    "GlobalAvgPoolCNN": build_globalavgpool_cnn((*img_size, 3), num_classes)
}

histories = {}
final_accuracies = {}

for name, model in models.items():
    print(f"\n=== Training {name} ===")
    hist = model.fit(train_gen, validation_data=val_gen, epochs=epochs, verbose=2)
    histories[name] = hist.history
    loss, acc = model.evaluate(val_gen, verbose=0)
    final_accuracies[name] = acc
    model.save(f"medical_mnist_{name.lower()}.keras")
    print(f"{name}: Final Validation Accuracy = {acc:.4f}")

# ----------------------------
# Validation accuracy plot
# ----------------------------
plt.figure(figsize=(10,6))
for name, hist in histories.items():
    plt.plot(hist["val_accuracy"], label=f"{name}")
plt.title("Validation Accuracy Across Epochs")
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# ----------------------------
# Confusion matrices
# ----------------------------
for name, model in models.items():
    print(f"\n=== Confusion Matrix: {name} ===")
    val_preds = np.argmax(model.predict(val_gen), axis=1)
    true_labels = val_gen.classes
    cm = confusion_matrix(true_labels, val_preds)
    acc = final_accuracies[name]
    plt.figure(figsize=(7,6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_labels, yticklabels=class_labels)
    plt.title(f"{name} Confusion Matrix (Acc={acc:.3f})")
    plt.ylabel("True Label")
    plt.xlabel("Predicted Label")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Model
import tensorflow as tf

def visualize_image_propagation(models, img, class_labels, max_filters=6):
    """
    Visualize how an input image propagates through the convolutional layers
    of multiple models. Each row corresponds to one model.
    Shows the input image clearly before the model visualization grid.
    """
    # --- Display the input image clearly first ---
    plt.figure(figsize=(4, 4))
    vis_img = img[0]
    vis_img = (vis_img - vis_img.min()) / (vis_img.max() - vis_img.min() + 1e-8)
    plt.imshow(vis_img, vmin=0, vmax=1)
    plt.title("Input Image")
    plt.axis("off")
    plt.show()

    # --- Then visualize model propagation ---
    max_layers = max(len([l for l in m.layers if isinstance(l, tf.keras.layers.Conv2D)]) for m in models.values())

    fig, axes = plt.subplots(len(models), max_layers + 1,
                             figsize=(4 * (max_layers + 1), 4 * len(models)))
    if len(models) == 1:
        axes = np.expand_dims(axes, axis=0)

    for m_idx, (name, model) in enumerate(models.items()):
        # Ensure model is built
        try:
            _ = model.predict(img, verbose=0)
        except Exception:
            if not hasattr(model, "inputs") or model.inputs is None:
                model.build(img.shape)
            _ = model.predict(img, verbose=0)

        # Predicted class
        preds = model.predict(img, verbose=0)
        pred_idx = np.argmax(preds[0])
        pred_label = class_labels[pred_idx]

        # Input tensor
        input_tensor = model.inputs if hasattr(model, "inputs") and model.inputs else model.layers[0].input

        # Select Conv2D layers
        conv_layers = [layer for layer in model.layers if isinstance(layer, tf.keras.layers.Conv2D)]
        n_conv = len(conv_layers)
        if not conv_layers:
            print(f"Skipping {name}  no Conv2D layers.")
            continue

        activation_model = Model(inputs=input_tensor, outputs=[l.output for l in conv_layers])
        activations = activation_model.predict(img, verbose=0)

        # Column 0: input image per model
        axes[m_idx, 0].imshow(vis_img, vmin=0, vmax=1)
        axes[m_idx, 0].set_title(f"{name}\nPredicted: {pred_label}", fontsize=11)
        axes[m_idx, 0].axis("off")

        # Columns 1..N: feature maps
        for l_idx, (layer, act) in enumerate(zip(conv_layers, activations), start=1):
            if act.ndim == 4:
                act_img = np.mean(act[0, :, :, :min(act.shape[-1], max_filters)], axis=-1)
            elif act.ndim == 3:
                act_img = np.mean(act[:, :, :min(act.shape[-1], max_filters)], axis=-1)
            else:
                continue

            axes[m_idx, l_idx].imshow(act_img, cmap="viridis")
            axes[m_idx, l_idx].set_title(layer.name, fontsize=9)
            axes[m_idx, l_idx].axis("off")

        # Hide unused columns for smaller models
        for j in range(n_conv + 1, max_layers + 1):
            axes[m_idx, j].remove()

    plt.tight_layout()
    plt.show()


val_imgs, val_labels = next(val_gen)
sample_img = val_imgs[0:1]
visualize_image_propagation(models, sample_img, class_labels)

val_imgs, val_labels = next(val_gen)
sample_img = val_imgs[2:3]
visualize_image_propagation(models, sample_img, class_labels)


```
