```python
# ==============================================================
# CNN (ResNet‑50) on FFT‑Images – RML2018.01A   (SNR > 3 dB)
# ==============================================================

# ---- 1. Imports ------------------------------------------------
import h5py, ast, numpy as np, matplotlib.pyplot as plt, seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
import tensorflow as tf

# ---- 2. Parameters / paths ------------------------------------
H5_PATH       = "RML2018/2018.01/GOLD_XYZ_OSC.0001_1024.hdf5"
CLASSES_TXT   = "RML2018/2018.01/classes.txt"
SNR_MIN       = 3        # keep only >3 dB
IMG_SIZE      = 64
MAX_PER_CLASS = 2000
EPOCHS        = 30
BATCH_SIZE    = 64
LR            = 1e-4

# ---- 3. Helper: load IQ data into per_cls ---------------------
def load_rml2018_to_dict(h5_path, classes_txt, snr_min=3):
    """Return per_cls[cls] -> list[(1024,3) float32] with SNR column."""
    with open(classes_txt, "r") as f:
        class_list = ast.literal_eval(f.read().split("=")[-1].strip())
        class_list = [c.strip(" '") for c in class_list]

    per_cls = {c: [] for c in class_list}
    with h5py.File(h5_path, "r") as f:
        X, Y, Z = f["X"][:], f["Y"][:], f["Z"][:]

    for i in range(len(X)):
        snr = int(Z[i][0])
        if snr <= snr_min:
            continue
        cls = class_list[int(Y[i].argmax())]
        sig = np.hstack([X[i], np.full((1024, 1), snr, dtype=np.float32)])
        if len(per_cls[cls]) < MAX_PER_CLASS:
            per_cls[cls].append(sig)
    return per_cls

# ---- 4. Helper: IQ→FFT‑image dataset --------------------------
def build_rf_image_dataset(per_cls, img_size=64):
    X_img, y = [], []
    for cls, samples in per_cls.items():
        for sig in samples:
            iq = sig[:, :2]
            cmplx = iq[:, 0] + 1j * iq[:, 1]
            fft_mag = np.abs(np.fft.fftshift(np.fft.fft(cmplx)))
            fft_mag = 20 * np.log10(fft_mag + 1e-6)
            fft_mag = (fft_mag - fft_mag.min()) / (fft_mag.max() - fft_mag.min() + 1e-9)

            # square image
            needed = img_size * img_size
            if fft_mag.size < needed:
                fft_mag = np.pad(fft_mag, (0, needed - fft_mag.size))
            img2d = fft_mag[:needed].reshape(img_size, img_size)
            img3d = np.repeat(img2d[..., None], 3, axis=2).astype(np.float32)
            X_img.append(img3d)
            y.append(cls)

    X_img = np.stack(X_img)
    le = LabelEncoder(); y_enc = le.fit_transform(y)
    return X_img, y_enc, le

# ---- 5. Build dataset ----------------------------------------
per_cls = load_rml2018_to_dict(H5_PATH, CLASSES_TXT, snr_min=SNR_MIN)
X_img, y_enc, le = build_rf_image_dataset(per_cls, IMG_SIZE)
print(f"Dataset: {X_img.shape[0]} images  |  classes: {le.classes_}")

# ---- 6. Train / test split & normalisation -------------------
X_img = X_img / 255.0
X_tr, X_te, y_tr, y_te = train_test_split(
    X_img, y_enc, stratify=y_enc, test_size=0.20, random_state=42
)
y_tr_cat, y_te_cat = to_categorical(y_tr), to_categorical(y_te)
num_classes = len(le.classes_)

# ---- 7. Build ResNet‑50 model --------------------------------
base = ResNet50(weights="imagenet", include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
x = GlobalAveragePooling2D()(base.output)
x = Dense(512, activation="relu")(x); x = Dropout(0.4)(x)
x = Dense(256, activation="relu")(x)
outputs = Dense(num_classes, activation="softmax")(x)
model = Model(inputs=base.input, outputs=outputs)
model.compile(optimizer=Adam(LR), loss="categorical_crossentropy", metrics=["accuracy"])
model.summary()

# ---- 8. Train -------------------------------------------------
history = model.fit(
    X_tr, y_tr_cat,
    validation_data=(X_te, y_te_cat),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    verbose=2
)

# ---- 9. Evaluate ---------------------------------------------
y_pred = np.argmax(model.predict(X_te, verbose=0), axis=1)
print("\nClassification Report:")
print(classification_report(y_te, y_pred, target_names=le.classes_))

cm = confusion_matrix(y_te, y_pred, normalize="true")
plt.figure(figsize=(12, 10))
sns.heatmap(cm, cmap="Blues", xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("Normalised Confusion Matrix – ResNet‑50 on RF FFT Images")
plt.xticks(rotation=90); plt.yticks(rotation=0); plt.tight_layout(); plt.show()
```

```python
# ==============================================================
#  TRAIN & SAVE  — fixed output path (no timestamp)
# ==============================================================

import h5py, ast, math, os, random, pickle, datetime
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam

# ------------------- CONFIG ------------------------------------
H5_PATH, CLASSES_TXT = "RML2018/2018.01/GOLD_XYZ_OSC.0001_1024.hdf5", "RML2018/2018.01/classes.txt"
SNR_CUTOFF   = 6
EXCLUDE      = {"AM-DSB", "WBFM"}
IMG_SIZE     = 64
MAX_PER_CLS  = 2000
CROP_FRAC    = 0.50
DECIM        = 2
N_FFT        = 1024
EPOCHS       = 30
BATCH_SIZE   = 64
LR           = 1e-4

OUT_DIR = f"saved_models/fft64_snrcut{SNR_CUTOFF}"
os.makedirs(OUT_DIR, exist_ok=True)

# ------------------- HELPERS -----------------------------------
def load_per_class(h5_path, cls_txt):
    with open(cls_txt) as f:
        names = [c.strip(" '") for c in ast.literal_eval(f.read().split("=")[-1].strip())]
    per_cls = {c: [] for c in names}
    with h5py.File(h5_path) as f:
        X, Y, Z = f["X"][:], f["Y"][:], f["Z"][:]
    for i in range(len(X)):
        cls  = names[int(Y[i].argmax())]
        per_cls[cls].append(
            np.hstack([X[i], np.full((1024, 1), int(Z[i][0]), np.float32)]))
    return per_cls

def build_buffer(per_cls, snr_min, max_per, exclude):
    buf = {}
    for c, sigs in per_cls.items():
        if c in exclude: continue
        sigs = [s for s in sigs if s[0, 2] > snr_min]
        if not sigs: continue
        random.shuffle(sigs)
        buf[c] = sigs[:max_per]
    return buf

def sig_to_img(sig, img_size, n_fft, crop_frac, decim):
    iq = sig[:, :2]; cplx = iq[:,0] + 1j*iq[:,1]
    fft = np.abs(np.fft.fftshift(np.fft.fft(cplx, n_fft)))
    mid, half = n_fft//2, int(n_fft*crop_frac/2)
    fft = fft[mid-half:mid+half:decim]
    fft = 20*np.log10(fft+1e-6)
    fft = (fft-fft.min())/(fft.max()-fft.min()+1e-9)
    need = img_size**2
    if fft.size < need:
        fft = np.tile(fft, -(-need//fft.size))[:need]
    img = fft[:need].reshape(img_size, img_size)
    return np.repeat(img[...,None], 3, 2).astype(np.float32)

def buffer_to_dataset(buf):
    X, y = [], []
    for c, sigs in buf.items():
        for s in sigs:
            X.append(sig_to_img(s, IMG_SIZE, N_FFT, CROP_FRAC, DECIM))
            y.append(c)
    X = np.stack(X)
    le = LabelEncoder(); y_enc = le.fit_transform(y)
    return X, y_enc, le

# ------------------- DATA --------------------------------------
per_cls_all = load_per_class(H5_PATH, CLASSES_TXT)
train_buf   = build_buffer(per_cls_all, SNR_CUTOFF, MAX_PER_CLS, EXCLUDE)
X, y, le    = buffer_to_dataset(train_buf)
X = X/255.0
X_tr, X_val, y_tr, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
y_tr, y_val = to_categorical(y_tr), to_categorical(y_val)

# ------------------- MODEL -------------------------------------
base = ResNet50(include_top=False, weights="imagenet",
                input_shape=(IMG_SIZE, IMG_SIZE, 3))
for l in base.layers[:-20]:
    l.trainable = False
x = GlobalAveragePooling2D()(base.output)
x = Dense(512, activation="relu")(x); x = Dropout(0.4)(x)
x = Dense(256, activation="relu")(x)
out = Dense(len(le.classes_), activation="softmax")(x)
model = Model(base.input, out)
model.compile(optimizer=Adam(LR), loss="categorical_crossentropy", metrics=["accuracy"])
model.fit(X_tr, y_tr, validation_data=(X_val, y_val),
          epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=2)

# ------------------- SAVE (fixed path) --------------------------
model.save(os.path.join(OUT_DIR, "saved_model.keras"), overwrite=True)
with open(os.path.join(OUT_DIR, "label_encoder.pkl"), "wb") as f:
    pickle.dump(le, f)
print("✓ model & encoder saved to", OUT_DIR)
```

```python
# =============================================================
# Open World Learning – RML2018a → 128x64 IQ+FFT Images (24 classes)
# =============================================================
import h5py, ast, numpy as np, matplotlib.pyplot as plt, os, io, random
from sklearn.metrics import confusion_matrix
from scipy.fft import fft
from scipy.stats import entropy
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from PIL import Image
import seaborn as sns
import tensorflow as tf

# ------------------------------------------
# 🔹 Composite Image Generator (IQ + FFT)
# ------------------------------------------
def generate_image(iq_data, snr):
    fig, axs = plt.subplots(1, 2, figsize=(2.56, 1.28), dpi=50)  # 128x64 px total

    # IQ plot
    axs[0].plot(iq_data[:,0], label='I', color='cyan', linewidth=0.5)
    axs[0].plot(iq_data[:,1], label='Q', color='magenta', linewidth=0.5)
    axs[0].axis("off")

    # FFT magnitude plot
    fft_vals = np.abs(fft(iq_data[:,0] + 1j*iq_data[:,1]))
    axs[1].plot(fft_vals[:len(fft_vals)//2], color='orange', linewidth=0.5)
    axs[1].axis("off")

    buf = io.BytesIO()
    plt.tight_layout(pad=0)
    plt.savefig(buf, format='png')
    plt.close(fig)
    buf.seek(0)
    img = Image.open(buf).convert('RGB').resize((128,64))
    buf.close()
    return np.asarray(img)/255.0

# ------------------------------------------
# 🔹 Load and preprocess RML2018a data
# ------------------------------------------
def load_rml2018a(h5_path, class_txt, snr_range=(4,30), max_per_class=2000):
    with open(class_txt, "r") as f:
        class_list = ast.literal_eval(f.read().split("=")[-1].strip())
    with h5py.File(h5_path, "r") as f:
        X_raw, Y_raw, Z_raw = f["X"][:], f["Y"][:], f["Z"][:]
    data = {cls: [] for cls in class_list}
    for i in range(len(X_raw)):
        snr = int(Z_raw[i][0])
        if snr_range[0] <= snr <= snr_range[1]:
            label = class_list[np.argmax(Y_raw[i])]
            if len(data[label]) < max_per_class:
                iq = X_raw[i]
                data[label].append((iq, snr))
    return data

# ------------------------------------------
# 🔹 Create dataset for given class subset
# ------------------------------------------
def make_dataset(data_dict, class_names):
    x, y = [], []
    for i, cls in enumerate(class_names):
        for iq, snr in data_dict[cls]:
            x.append(generate_image(iq, snr))
            y.append(i)
    x = np.array(x, dtype=np.float32)
    y = to_categorical(y, num_classes=len(class_names))
    return x, y

# ------------------------------------------
# 🔹 Unknown Detection via Softmax + Entropy
# ------------------------------------------
def detect_unknowns(model, x_test, thresh_softmax=0.8, thresh_entropy=1.0):
    probs = model.predict(x_test)
    max_conf = np.max(probs, axis=1)
    ent = entropy(probs.T)
    return np.where((max_conf < thresh_softmax) | (ent > thresh_entropy))[0]

# ------------------------------------------
# 🔹 Build classifier
# ------------------------------------------
def build_model(num_classes):
    base = ResNet50(weights="imagenet", include_top=False, input_shape=(64,128,3))
    x = GlobalAveragePooling2D()(base.output)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.3)(x)
    x = Dense(256, activation='relu')(x)
    out = Dense(num_classes, activation='softmax')(x)
    model = Model(base.input, out)
    model.compile(optimizer=Adam(1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# ------------------------------------------
# 🔹 Main open world loop
# ------------------------------------------
def train_open_world():
    # Load data
    data = load_rml2018a("RML2018/2018.01/GOLD_XYZ_OSC.0001_1024.hdf5",
                         "RML2018/2018.01/classes.txt")

    all_classes = sorted(data.keys())
    initial_classes = all_classes[:10]
    remaining_classes = all_classes[10:]

    print(f"Known: {initial_classes}")
    print(f"Remaining: {remaining_classes}")

    model = build_model(len(initial_classes))
    x_train, y_train = make_dataset(data, initial_classes)
    x_test, y_test = x_train.copy(), y_train.copy()  # simple split

    print("\n🔧 Training initial 10-class model...")
    model.fit(x_train, y_train, validation_split=0.1, epochs=10, batch_size=64)

    learned = initial_classes.copy()
    while remaining_classes:
        batch = remaining_classes[:5] if len(remaining_classes) >= 5 else remaining_classes
        print(f"\n📌 Checking for unknowns... (next={batch})")
        unknown_idx = detect_unknowns(model, x_test)

        if len(unknown_idx) > 200:
            print(f"➡️ Detected {len(unknown_idx)} unknowns. Expanding with new classes: {batch}")
            learned += batch
            remaining_classes = remaining_classes[len(batch):]

            # Regenerate data
            x_train, y_train = make_dataset(data, learned)
            x_test, y_test = x_train.copy(), y_train.copy()

            # Expand model
            model = build_model(len(learned))
            model.fit(x_train, y_train, validation_split=0.1, epochs=10, batch_size=64)
        else:
            print("↪️  Not enough unknowns confidently detected. Skipping expansion.")
            break

    return model, x_test, y_test, learned

# -------------------------
# 🔹 Execute training loop
# -------------------------
final_model, final_x, final_y, final_classes = train_open_world()

# Evaluate
pred = np.argmax(final_model.predict(final_x), axis=1)
true = np.argmax(final_y, axis=1)
cm = confusion_matrix(true, pred)

plt.figure(figsize=(12,10))
sns.heatmap(cm, fmt='d', cmap='Blues')
plt.title("Final Confusion Matrix – RML2018a")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.show()
```

```python
# =============================================================
# Final Replay Fine-Tuning of RML2018a Model (24 Classes)
# =============================================================
from tensorflow.keras.callbacks import ReduceLROnPlateau

# ----------------------------------------
# 🔁 Replay Buffer
# ----------------------------------------
class ReplayBuffer:
    def __init__(self, capacity=5000):
        self.capacity = capacity
        self.buffer = []

    def add(self, x, y):
        for i in range(len(x)):
            self.buffer.append((x[i], y[i]))
        if len(self.buffer) > self.capacity:
            self.buffer = self.buffer[-self.capacity:]

    def sample(self, n):
        idxs = np.random.choice(len(self.buffer), size=n, replace=False)
        xs, ys = zip(*[self.buffer[i] for i in idxs])
        return np.array(xs), np.array(ys)

# ----------------------------------------
# 🔄 Rebuild model with soft labels
# ----------------------------------------
def build_refined_model(num_classes):
    base = ResNet50(weights="imagenet", include_top=False, input_shape=(64,128,3))
    x = GlobalAveragePooling2D()(base.output)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.3)(x)
    x = Dense(256, activation='relu')(x)
    out = Dense(num_classes, activation='softmax')(x)
    model = Model(base.input, out)
    model.compile(optimizer=Adam(1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# ----------------------------------------
# 🧪 Final 24-class training
# ----------------------------------------
def final_refinement(full_data, class_list, epochs=30, batch_size=64):
    x_all, y_all = make_dataset(full_data, class_list)
    buffer = ReplayBuffer()
    buffer.add(x_all, y_all)

    model = build_refined_model(len(class_list))
    lr_callback = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1)

    print(f"\n🔁 Final fine-tuning on all {len(class_list)} classes...")
    history = model.fit(
        x_all, y_all,
        validation_split=0.1,
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[lr_callback],
        verbose=1
    )
    return model, x_all, y_all

# ----------------------------------------
# 🔍 Evaluate refined model
# ----------------------------------------
refined_model, x_all_final, y_all_final = final_refinement(data, final_classes)

y_pred = np.argmax(refined_model.predict(x_all_final), axis=1)
y_true = np.argmax(y_all_final, axis=1)
cm = confusion_matrix(y_true, y_pred)

# ----------------------------------------
# 🔳 Normalized Confusion Matrix
# ----------------------------------------
cm_norm = cm.astype(np.float32) / cm.sum(axis=1, keepdims=True)

plt.figure(figsize=(13,10))
sns.heatmap(cm_norm, cmap="Blues", xticklabels=final_classes, yticklabels=final_classes)
plt.title("Final Normalized Confusion Matrix – RML2018a (24 Classes)")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.tight_layout(); plt.show()
```
